```python
def word_frequency(input_str):
    """
    Compte la fréquence d'apparition de chaque mot unique dans une chaîne.

    Args:
        input_str (str): La chaîne de caractères à analyser

    Returns:
        dict: Dictionnaire avec les mots comme clés et leurs fréquences comme valeurs

    Time Complexity: O(n) où n est le nombre total de caractères
    Space Complexity: O(k) où k est le nombre de mots uniques
    """
    # Cas d'une chaîne vide
    if not input_str or not input_str.strip():
        return {}

    # Nettoyer la chaîne : supprimer la ponctuation et mettre en minuscules
    # Pour une version plus robuste, on peut utiliser une expression régulière
    import re
    # Remplacer la ponctuation par des espaces et diviser
    cleaned_str = re.sub(r'[^\w\s]', ' ', input_str.lower())
    words = cleaned_str.split()

    # Dictionnaire pour stocker les fréquences
    frequency_dict = {}

    # Parcourir chaque mot et incrémenter son compteur
    for word in words:
        frequency_dict[word] = frequency_dict.get(word, 0) + 1

    return frequency_dict

# Version alternative plus simple (sans regex)
def word_frequency_simple(input_str):
    """
    Version simplifiée sans regex pour une meilleure compréhension.
    """
    if not input_str or not input_str.strip():
        return {}

    # Mettre en minuscules et diviser par les espaces
    words = input_str.lower().split()
    frequency_dict = {}

    for word in words:
        # Supprimer la ponctuation des mots (uniquement la ponctuation aux extrémités)
        word = word.strip('.,!?;:()[]{}"\'')
        if word:  # Ignorer les mots vides (après suppression de la ponctuation)
            frequency_dict[word] = frequency_dict.get(word, 0) + 1

    return frequency_dict

# Version avancée avec gestion des stop words
def word_frequency_advanced(input_str, remove_stop_words=False):
    """
    Version avancée avec options supplémentaires.

    Args:
        input_str: La chaîne à analyser
        remove_stop_words: Si True, supprime les mots courants (le, la, les, etc.)
    """
    if not input_str or not input_str.strip():
        return {}

    import re

    # Nettoyer et séparer
    cleaned_str = re.sub(r'[^\w\s]', ' ', input_str.lower())
    words = cleaned_str.split()

    # Mots vides (stop words) en français
    stop_words = {'le', 'la', 'les', 'un', 'une', 'des', 'et', 'ou', 'mais',
                  'donc', 'or', 'ni', 'car', 'en', 'dans', 'sur', 'sous', 'pour',
                  'avec', 'sans', 'par', 'à', 'de', 'du', 'au', 'aux', 'est',
                  'sont', 'était', 'étaient', 'a', 'ont', 'avait', 'avaient'}

    frequency_dict = {}

    for word in words:
        # Nettoyer la ponctuation
        word = word.strip('.,!?;:()[]{}"\'')
        if not word:  # Ignorer les mots vides
            continue

        # Optionnel : ignorer les stop words
        if remove_stop_words and word in stop_words:
            continue

        frequency_dict[word] = frequency_dict.get(word, 0) + 1

    return frequency_dict

# Fonctions de test
def test_word_frequency():
    """Teste la fonction word_frequency avec différents cas."""

    print("=" * 70)
    print("📊 DÉNOMBREMENT DES FRÉQUENCES - TESTS")
    print("=" * 70)

    test_cases = [
        ("apple orange apple banana orange apple", {'apple': 3, 'orange': 2, 'banana': 1}),
        ("hello world hello", {'hello': 2, 'world': 1}),
        ("the cat in the hat", {'the': 2, 'cat': 1, 'in': 1, 'hat': 1}),
        ("one two three four five", {'one': 1, 'two': 1, 'three': 1, 'four': 1, 'five': 1}),
        ("", {}),
        ("   ", {}),
        ("word word word word", {'word': 4}),
        ("Bonjour! Comment ça va? Bonjour, ça va bien!", {'bonjour': 2, 'comment': 1, 'ça': 2, 'va': 2, 'bien': 1}),
    ]

    for i, (input_str, expected) in enumerate(test_cases, 1):
        result = word_frequency(input_str)
        status = "✅" if result == expected else "❌"
        print(f"\nTest {i}: {status}")
        print(f"  Input: '{input_str}'")
        print(f"  Résultat: {result}")
        print(f"  Attendu: {expected}")

# Exemple d'utilisation interactive
def interactive_word_frequency():
    """Mode interactif pour tester la fonction."""

    print("\n" + "=" * 70)
    print("🎮 MODE INTERACTIF - DÉNOMBREMENT DES FRÉQUENCES")
    print("=" * 70)
    print("Entrez une phrase pour analyser la fréquence des mots")
    print("Entrez 'quit' pour quitter")
    print("-" * 70)

    while True:
        user_input = input("\n📝 Entrez une phrase: ").strip()

        if user_input.lower() == 'quit':
            print("👋 Au revoir !")
            break

        if not user_input:
            print("⚠️ Veuillez entrer une phrase valide.")
            continue

        # Analyser la phrase
        frequencies = word_frequency(user_input)

        print("\n📊 Résultats:")
        print("-" * 40)
        if frequencies:
            # Trier par fréquence décroissante
            sorted_freq = sorted(frequencies.items(), key=lambda x: x[1], reverse=True)
            for word, count in sorted_freq:
                print(f"  {word}: {count}")
        else:
            print("  Aucun mot trouvé.")
        print("-" * 40)

if __name__ == "__main__":
    # Exécution des tests
    test_word_frequency()

    # Exemple de l'énoncé
    print("\n" + "=" * 70)
    print("📘 EXEMPLE DE L'ÉNONCÉ")
    print("=" * 70)
    input_str = "apple orange apple banana orange apple"
    result = word_frequency(input_str)
    print(f"Input: '{input_str}'")
    print(f"Output: {result}")

    # Mode interactif (décommentez pour utiliser)
    # interactive_word_frequency()
```

```python
def longest_unique_substring(s):
    """
    Trouve la longueur de la plus longue sous-chaîne sans caractères répétés.

    Args:
        s (str): La chaîne de caractères à analyser

    Returns:
        int: La longueur de la plus longue sous-chaîne sans répétition

    Time Complexity: O(n) où n est la longueur de la chaîne
    Space Complexity: O(min(n, m)) où m est la taille de l'alphabet
    """
    if not s:
        return 0

    # Dictionnaire pour stocker le dernier index de chaque caractère
    last_seen = {}

    # Variables pour la fenêtre glissante
    start = 0  # Début de la fenêtre actuelle
    max_length = 0  # Longueur maximale trouvée

    # Parcourir la chaîne avec un index 'end'
    for end, char in enumerate(s):
        # Si le caractère est déjà dans la fenêtre actuelle
        if char in last_seen and last_seen[char] >= start:
            # Déplacer le début de la fenêtre après la dernière occurrence
            start = last_seen[char] + 1

        # Mettre à jour la dernière occurrence du caractère
        last_seen[char] = end

        # Mettre à jour la longueur maximale
        current_length = end - start + 1
        max_length = max(max_length, current_length)

    return max_length

# Version qui retourne aussi la sous-chaîne
def longest_unique_substring_with_string(s):
    """
    Version améliorée qui retourne à la fois la longueur et la sous-chaîne.

    Returns:
        tuple: (longueur, sous-chaîne)
    """
    if not s:
        return 0, ""

    last_seen = {}
    start = 0
    max_length = 0
    max_start = 0  # Début de la sous-chaîne maximale

    for end, char in enumerate(s):
        if char in last_seen and last_seen[char] >= start:
            start = last_seen[char] + 1

        last_seen[char] = end

        current_length = end - start + 1
        if current_length > max_length:
            max_length = current_length
            max_start = start

    return max_length, s[max_start:max_start + max_length]

# Version utilisant un ensemble (Set) - alternative
def longest_unique_substring_set(s):
    """
    Version utilisant un ensemble au lieu d'un dictionnaire.
    Cette version est plus intuitive mais légèrement moins efficace pour
    certains cas extrêmes car elle doit retirer des éléments de l'ensemble.
    """
    if not s:
        return 0

    char_set = set()
    start = 0
    max_length = 0

    for end, char in enumerate(s):
        # Tant que le caractère est déjà dans l'ensemble
        while char in char_set:
            # Retirer le caractère au début de la fenêtre
            char_set.remove(s[start])
            start += 1

        # Ajouter le nouveau caractère
        char_set.add(char)
        max_length = max(max_length, end - start + 1)

    return max_length

# Version avec validation de sous-chaîne (pour debug)
def longest_unique_substring_debug(s):
    """
    Version avec affichage des étapes pour comprendre l'algorithme.
    """
    if not s:
        return 0

    print(f"Analyse de: '{s}'")
    print("-" * 50)

    last_seen = {}
    start = 0
    max_length = 0
    current_substring = ""

    for end, char in enumerate(s):
        if char in last_seen and last_seen[char] >= start:
            start = last_seen[char] + 1
            print(f"  ⚡ Répétition de '{char}' -> déplacement du début à {start}")

        last_seen[char] = end
        current_substring = s[start:end+1]
        current_length = end - start + 1

        print(f"  end={end}, char='{char}', fenêtre='{current_substring}', longueur={current_length}")

        if current_length > max_length:
            max_length = current_length
            print(f"  🏆 Nouvelle longueur max: {max_length}")

    print(f"✅ Résultat: {max_length}")
    return max_length

# Fonctions de test
def test_longest_substring():
    """Teste la fonction avec différents cas."""

    print("=" * 70)
    print("🔍 SOUS-CHAÎNE LA PLUS LONGUE SANS RÉPÉTITION - TESTS")
    print("=" * 70)

    test_cases = [
        ("abcabcbb", 3, "abc"),
        ("bbbbb", 1, "b"),
        ("pwwkew", 3, "wke"),
        ("", 0, ""),
        ("a", 1, "a"),
        ("ab", 2, "ab"),
        ("aba", 2, "ab"),
        ("abc", 3, "abc"),
        ("dvdf", 3, "vdf"),
        ("tmmzuxt", 5, "mzuxt"),
        ("abcdabc", 4, "abcd"),
        ("aabaab!bb", 3, "ab!"),
    ]

    for i, (s, expected_length, expected_substring) in enumerate(test_cases, 1):
        result = longest_unique_substring(s)
        result_str, substr = longest_unique_substring_with_string(s)

        status = "✅" if result == expected_length else "❌"
        print(f"\nTest {i}: {status}")
        print(f"  Input: '{s}'")
        print(f"  Longueur: {result} (attendu: {expected_length})")
        print(f"  Sous-chaîne: '{substr}' (attendu: '{expected_substring}')")

def interactive_longest_substring():
    """Mode interactif pour tester la fonction."""

    print("\n" + "=" * 70)
    print("🎮 MODE INTERACTIF - SOUS-CHAÎNE SANS RÉPÉTITION")
    print("=" * 70)
    print("Entrez une chaîne pour trouver la plus longue sous-chaîne sans répétition")
    print("Entrez 'quit' pour quitter")
    print("-" * 70)

    while True:
        user_input = input("\n📝 Entrez une chaîne: ").strip()

        if user_input.lower() == 'quit':
            print("👋 Au revoir !")
            break

        if not user_input:
            print("⚠️ Veuillez entrer une chaîne valide.")
            continue

        # Analyse complète
        length, substring = longest_unique_substring_with_string(user_input)

        print("\n📊 Résultats:")
        print("-" * 50)
        print(f"  Chaîne: '{user_input}'")
        print(f"  Longueur maximale: {length}")
        print(f"  Sous-chaîne: '{substring}'")

        # Afficher la sous-chaîne en surbrillance dans la chaîne originale
        if substring:
            print(f"  Position: {user_input.find(substring)}")
        print("-" * 50)

if __name__ == "__main__":
    # Exécution des tests
    test_longest_substring()

    # Exemples de l'énoncé
    print("\n" + "=" * 70)
    print("📘 EXEMPLES DE L'ÉNONCÉ")
    print("=" * 70)

    examples = [
        ("abcabcbb", 3),
        ("bbbbb", 1),
        ("pwwkew", 3),
    ]

    for s, expected in examples:
        result = longest_unique_substring(s)
        _, substr = longest_unique_substring_with_string(s)
        print(f"\nInput: s = \"{s}\"")
        print(f"Output: {result}")
        print(f"Explanation: The answer is \"{substr}\", with the length of {result}.")
        print(f"Status: {'✅' if result == expected else '❌'}")

    # Démonstration détaillée
    print("\n" + "=" * 70)
    print("🔬 DÉMONSTRATION DÉTAILLÉE")
    print("=" * 70)
    longest_unique_substring_debug("abcabcbb")

    # Mode interactif (décommentez pour utiliser)
    # interactive_longest_substring()
```
