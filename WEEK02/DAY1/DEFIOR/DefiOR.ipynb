```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from scipy.ndimage import rotate
import requests
from io import BytesIO
import zipfile
import os

url = "https://github.com/devtlv/Datasets-DA-Bootcamp-2-/raw/refs/heads/main/Week%204%20-%20Data%20Understanding/W4D3%20-%20Importing%20Data%2C%20Exporting%20D/Flower%20Color%20Images%20Dataset.zip"

print("⏳ Téléchargement en cours...")
response = requests.get(url)
with zipfile.ZipFile(BytesIO(response.content)) as z:
    z.extractall("flowers")
print("✅ Dataset extrait !")

image_path = None
for root, dirs, files in os.walk("flowers"):
    for f in files:
        if f.endswith(".png") or f.endswith(".jpg"):
            image_path = os.path.join(root, f)
            break
    if image_path:
        break

print(f"📸 Image trouvée : {image_path}")

original_image = Image.open(image_path)

plt.figure(figsize=(5, 5))
plt.imshow(original_image)
plt.title("Image Originale")
plt.axis("off")
plt.show()

def rotate_image_30_degrees(image):
    image_array = np.array(image)
    rotated_array = rotate(image_array, angle=30, reshape=False)
    return Image.fromarray(rotated_array.astype(np.uint8))

rotated_image = rotate_image_30_degrees(original_image)

plt.figure(figsize=(5, 5))
plt.imshow(rotated_image)
plt.title("Rotation 30°")
plt.axis("off")
plt.show()

flipped_horizontal = ImageOps.mirror(original_image)   # horizontal
flipped_vertical   = ImageOps.flip(original_image)     # vertical

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(flipped_horizontal)
axes[0].set_title("Retournement Horizontal")
axes[0].axis("off")

axes[1].imshow(flipped_vertical)
axes[1].set_title("Retournement Vertical")
axes[1].axis("off")

plt.show()

width, height = original_image.size
new_width  = int(width * 1.2)
new_height = int(height * 1.2)

zoomed_image = original_image.resize((new_width, new_height))

plt.figure(figsize=(5, 5))
plt.imshow(zoomed_image)
plt.title(f"Zoom x1.2 ({new_width}×{new_height}px)")
plt.axis("off")
plt.show()


fig, axes = plt.subplots(1, 5, figsize=(20, 4))

images = [original_image, rotated_image,
          flipped_horizontal, flipped_vertical, zoomed_image]

titres = ["Originale", "Rotation 30°",
          "Flip Horizontal", "Flip Vertical", "Zoom x1.2"]

for ax, img, titre in zip(axes, images, titres):
    ax.imshow(img)
    ax.set_title(titre)
    ax.axis("off")

plt.suptitle("Transformations d'Images — Flower Dataset", fontsize=14)
plt.tight_layout()
plt.show()

print("✅ Toutes les transformations sont terminées !")
```
