```python
#DAILY CHALLENGE

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("="*50)
print("DÉFI QUOTIDIEN : VISUALISATION")
print("="*50)

print("\n1. CHARGEMENT DES DONNÉES")

df = pd.read_excel(r'C:\Users\kmeso\Desktop\TTA_KOUAME_ASSE_SOKRY_JOSEPH\WEEK02\CVS\CVS3\US Superstore data.xls')

df = df.dropna()
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year

print(f"✓ {len(df)} lignes chargées")
print(df[['Sales', 'Profit', 'Discount']].head(3))

print("\n" + "="*40)
print("2. GRAPHIQUES AVEC MATPLOTLIB")
print("="*40)

print("\n Ligne : Ventes par année")
ventes_annee = df.groupby('Year')['Sales'].sum()

plt.figure(figsize=(10,5))
plt.plot(ventes_annee.index, ventes_annee.values, 'o-', linewidth=2, markersize=8, color='blue')
plt.fill_between(ventes_annee.index, ventes_annee.values, alpha=0.3)
plt.title('Ventes par Année', fontsize=14)
plt.xlabel('Année')
plt.ylabel('Ventes ($)')
plt.grid(True, alpha=0.3)

for x,y in zip(ventes_annee.index, ventes_annee.values):
    plt.text(x, y+5000, f'${y:,.0f}', ha='center')

plt.tight_layout()
plt.show()

print("\n Top États par Ventes")
ventes_etat = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(ventes_etat.index, ventes_etat.values, color='steelblue')
plt.title('Top 10 États par Ventes', fontsize=14)
plt.xlabel('Ventes ($)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

print(f"Top état : {ventes_etat.index[0]} (${ventes_etat.values[0]:,.0f})")

print("\n" + "="*40)
print("3. GRAPHIQUES AVEC SEABORN")
print("="*40)

print("\n Top 10 produits les plus vendus")
top_produits = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_produits.values, y=top_produits.index, palette='viridis')
plt.title('Top 10 Produits par Ventes', fontsize=14)
plt.xlabel('Ventes ($)')
plt.tight_layout()
plt.show()

print("\n Relation Profit vs Remise")
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category', alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--', alpha=0.7)
plt.title('Profit vs Remise par Catégorie', fontsize=14)
plt.tight_layout()
plt.show()

print("\n" + "="*40)
print("4. COMPARAISON MATPLOTLIB vs SEABORN")
print("="*40)

print("\n" + "="*40)
print("5. OBSERVATIONS")
print("="*40)

print("""
Ce qu'on apprend :

1. VENTES : Augmentation chaque année
2. ÉTATS : Californie, New York, Texas = meilleurs
3. PRODUITS : Quelques produits stars dominent
4. REMISES : Les grosses remises réduisent les profits
5. CATÉGORIES : Technology plus résistante aux remises
""")

print("="*50)
print("✓ DÉFI TERMINÉ")
print("="*50)
```
