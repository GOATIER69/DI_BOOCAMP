```python
#DAILY CHALLENGE

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
df = pd.DataFrame({
    'Date': pd.date_range('2020-01-01', '2023-12-31', periods=500),
    'Category': np.random.choice(['A', 'B', 'C'], 500),
    'Product': np.random.choice(['P1', 'P2', 'P3'], 500),
    'State': np.random.choice(['CA', 'NY', 'TX'], 500),
    'Sales': np.random.uniform(50, 500, 500),
    'Discount': np.random.uniform(0, 0.3, 500),
    'Profit': np.random.uniform(-30, 100, 500)
})

plt.figure(figsize=(10, 4))
plt.plot(df.groupby(df['Date'].dt.to_period('M'))['Sales'].sum().values, marker='o', color='steelblue')
plt.title('Tendance des ventes')
plt.ylabel('Ventes ($)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
df.groupby('State')['Sales'].sum().sort_values().plot(kind='barh', color='coral', edgecolor='black')
plt.title('Ventes par État')
plt.xlabel('Ventes ($)')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
top = df.groupby('Product')['Sales'].sum().nlargest(5)
sns.barplot(x=top.values, y=top.index, palette='viridis')
plt.title('Top 5 produits')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category', alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Remise vs Profit')
plt.tight_layout()
plt.show()

print("\n RÉSUMÉ:")
print(f"Ventes: ${df['Sales'].sum():,.0f} | Profit: ${df['Profit'].sum():,.0f}")
print(f"Meilleur état: {df.groupby('State')['Sales'].sum().idxmax()}")
print("Analyse terminée - Matplotlib pour les tendances, Seaborn pour les relations")
```
