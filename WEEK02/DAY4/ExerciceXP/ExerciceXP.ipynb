1. Importance de la visualisation des données :

La visualisation transforme des données brutes en images compréhensibles. Elle permet de détecter rapidement des tendances, des valeurs aberrantes et des relations entre variables. Elle facilite aussi la prise de décision et la communication avec des interlocuteurs non techniques.

2. Utilité d’un graphique linéaire :

Le graphique linéaire sert principalement à représenter l’évolution d’une ou plusieurs variables sur un axe continu (souvent le temps). Il est très utile pour suivre des tendances, comparer des séries et identifier des cycles ou des ruptures.

```python
#EXERCICE 2

import matplotlib.pyplot as plt

jours = ["Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi", "Samedi", "Dimanche"]
temperatures = [72, 74, 76, 80, 82, 78, 75]

plt.plot(jours, temperatures, marker="o", linestyle="-", color="red", linewidth=2)

plt.xlabel("Jour")
plt.ylabel("Température (°F)")
plt.title("Variation des températures sur une semaine")

plt.grid(True, linestyle="--", alpha=0.5) 
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(jours, temperatures, marker="s", linestyle="--", color="green", linewidth=2, markersize=8)
plt.xlabel("Jour", fontsize=12)
plt.ylabel("Température (°F)", fontsize=12)
plt.title("Variation des températures sur une semaine", fontsize=14, fontweight="bold")
plt.ylim(70, 85)
plt.grid(True, linestyle=":", alpha=0.7)
plt.show()
```

```python
#EXERCICE 3

import matplotlib.pyplot as plt

mois = ["Janvier", "Février", "Mars", "Avril", "Mai"]
ventes = [5000, 5500, 6200, 7000, 7500]

plt.bar(mois, ventes, color="skyblue", edgecolor="navy", linewidth=1.5)

plt.xlabel("Mois")
plt.ylabel("Montant des ventes ($)")
plt.title("Ventes mensuelles du magasin de détail")

for i, v in enumerate(ventes):
    plt.text(i, v + 100, f"${v}", ha="center", fontsize=10, fontweight="bold")

plt.show()
```

```python
#EXERCICE 4

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)  
cgpa_values = np.random.normal(loc=3.0, scale=0.5, size=200)  
cgpa_values = np.clip(cgpa_values, 1.5, 4.0)  

df = pd.DataFrame({'CGPA': cgpa_values})

plt.figure(figsize=(10, 6))
plt.hist(df['CGPA'], bins=15, color='steelblue', edgecolor='black', alpha=0.7)

plt.title('Distribution des moyennes cumulatives (CGPA) des étudiants', fontsize=14, fontweight='bold')
plt.xlabel('CGPA', fontsize=12)
plt.ylabel("Nombre d'étudiants", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```

```python
#EXERCICE 5

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\kmeso\Desktop\TTA_KOUAME_ASSE_SOKRY_JOSEPH\WEEK02\CVS\CVS3\Student Mental health.csv')

sex_column = 'Choose your gender' 
anxiety_column = 'Do you have Anxiety?' 

cross_tab = pd.crosstab(df[sex_column], df[anxiety_column])

plt.figure(figsize=(10, 6))

x = range(len(cross_tab.index))
width = 0.35

colors = ['#FF6B6B', '#4ECDC4']
for i, anxiety_level in enumerate(cross_tab.columns):
    plt.bar([pos + i*width for pos in x], cross_tab[anxiety_level].values, 
            width, label=anxiety_level, color=colors[i], edgecolor='black', alpha=0.8)

plt.xlabel('Sexe', fontsize=12, fontweight='bold')
plt.ylabel("Nombre d'étudiants", fontsize=12, fontweight='bold')
plt.title('Comparaison des niveaux d\'anxiété selon le sexe', fontsize=14, fontweight='bold')
plt.xticks([pos + width/2 for pos in x], cross_tab.index)
plt.legend(title='Niveau d\'anxiété')
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()

print("\nRésultats:")
print(cross_tab)
```

```python
#EXERCICE 6

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\kmeso\Desktop\TTA_KOUAME_ASSE_SOKRY_JOSEPH\WEEK02\CVS\CVS3\Student Mental health.csv')

age_column = 'Age' 
panic_column = 'Do you have Panic attack?' 

panic_mapping = {'Yes': 1, 'No': 0, 'yes': 1, 'no': 0, 'Oui': 1, 'Non': 0}
df['Panic_Numeric'] = df[panic_column].map(panic_mapping)

df['Age'] = pd.to_numeric(df[age_column], errors='coerce')
df_clean = df[['Age', 'Panic_Numeric']].dropna()

plt.figure(figsize=(12, 7))

non_panic = df_clean[df_clean['Panic_Numeric'] == 0]
panic = df_clean[df_clean['Panic_Numeric'] == 1]

jitter = np.random.uniform(-0.1, 0.1, len(non_panic))
plt.scatter(non_panic['Age'], non_panic['Panic_Numeric'] + jitter, 
            alpha=0.5, s=40, c='steelblue', label='Pas de crises')

jitter = np.random.uniform(-0.1, 0.1, len(panic))
plt.scatter(panic['Age'], panic['Panic_Numeric'] + jitter, 
            alpha=0.7, s=60, c='coral', label='Crises de panique')

plt.xlabel('Âge des étudiants', fontsize=12, fontweight='bold')
plt.ylabel('Crises de panique', fontsize=12, fontweight='bold')
plt.title('Relation entre l\'âge et les crises de panique', fontsize=14, fontweight='bold')
plt.yticks([0, 1], ['Non', 'Oui'])
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend()

from scipy import stats
z = np.polyfit(df_clean['Age'], df_clean['Panic_Numeric'], 1)
p = np.poly1d(z)
plt.plot(df_clean['Age'].sort_values(), p(df_clean['Age'].sort_values()), 
         "r--", linewidth=2, label='Tendance linéaire')

plt.tight_layout()
plt.show()

print(f"\nÂge moyen avec crises: {panic['Age'].mean():.2f} ans")
print(f"Âge moyen sans crises: {non_panic['Age'].mean():.2f} ans")
```
