```python
#EXERCICE 2

import numpy as np
from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]

moyenne    = np.mean(data)
mediane    = np.median(data)
variance   = np.var(data)
ecart_type = np.std(data)
mode       = stats.mode(data)

print(f"Données    : {data}")
print(f"Moyenne    : {moyenne:.2f}")
print(f"Médiane    : {mediane:.2f}")
print(f"Variance   : {variance:.2f}")
print(f"Écart-type : {ecart_type:.2f}")
print(f"Mode       : {mode.mode} (apparaît {mode.count} fois)")
```

```python
#EXERCICE 3

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

moyenne    = 50
ecart_type = 10

x    = np.linspace(20, 80, 300)
pdf  = norm.pdf(x, moyenne, ecart_type)  
data = norm.rvs(loc=moyenne, scale=ecart_type, size=1000, random_state=42)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(x, pdf, color='steelblue', linewidth=2)
axes[0].fill_between(x, pdf, alpha=0.3, color='steelblue')
axes[0].axvline(moyenne, color='red', linestyle='--', label=f'Moyenne = {moyenne}')
axes[0].axvline(moyenne + ecart_type, color='orange', linestyle='--', label=f'+1σ = {moyenne + ecart_type}')
axes[0].axvline(moyenne - ecart_type, color='orange', linestyle='--', label=f'-1σ = {moyenne - ecart_type}')
axes[0].set_title("Distribution Normale — Courbe PDF")
axes[0].set_xlabel("Valeurs")
axes[0].set_ylabel("Densité de probabilité")
axes[0].legend()

axes[1].hist(data, bins=30, density=True, color='steelblue',
             edgecolor='white', alpha=0.7, label='Données générées')
axes[1].plot(x, pdf, color='red', linewidth=2, label='Courbe théorique')
axes[1].set_title("Histogramme — 1000 échantillons")
axes[1].set_xlabel("Valeurs")
axes[1].set_ylabel("Densité")
axes[1].legend()

plt.suptitle("Distribution Normale (μ=50, σ=10)", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print(f"Moyenne théorique  : {moyenne}")
print(f"Moyenne générée    : {data.mean():.2f}")
print(f"Écart-type théo.   : {ecart_type}")
print(f"Écart-type généré  : {data.std():.2f}")
print(f"P(40 < X < 60)     : {norm.cdf(60, moyenne, ecart_type) - norm.cdf(40, moyenne, ecart_type):.2%}")
```

```python
#EXERCICE 4

import numpy as np
from scipy import stats

np.random.seed(42)
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

t_stat, p_value = stats.ttest_ind(data1, data2)

print(f"Moyenne data1  : {data1.mean():.2f}")
print(f"Moyenne data2  : {data2.mean():.2f}")
print(f"T-statistique  : {t_stat:.4f}")
print(f"P-value        : {p_value:.6f}")

alpha = 0.05
print(f"\nSeuil alpha    : {alpha}")
if p_value < alpha:
    print("✅ On rejette H0 → différence significative entre les deux groupes")
else:
    print("❌ On ne rejette pas H0 → pas de différence significative")
```

```python
#EXERCICE 5

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Données
house_sizes  = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

# Régression linéaire
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

print(f"Pente (slope)       : {slope:.2f}")
print(f"Ordonnée (intercept): {intercept:.2f}")
print(f"R²                  : {r_value**2:.4f}")
print(f"P-value             : {p_value:.6f}")
print(f"Erreur standard     : {std_err:.2f}")
#pour 90 m²
taille  = 90
prix_predit = slope * taille + intercept
print(f"\nPrix prédit pour {taille} m² : {prix_predit:,.0f}")

#Visuel
x_line = np.linspace(40, 130, 300)
y_line = slope * x_line + intercept

plt.figure(figsize=(10, 6))
plt.scatter(house_sizes, house_prices, color='steelblue',
            s=100, zorder=5, label='Données réelles')
plt.plot(x_line, y_line, color='red',
         linewidth=2, label='Droite de régression')
plt.scatter(taille, prix_predit, color='green',
            s=150, zorder=6, marker='*',
            label=f'Prédiction 90m² = {prix_predit:,.0f}')

for x, y in zip(house_sizes, house_prices):
    plt.annotate(f'{y:,}', (x, y),
                 textcoords="offset points",
                 xytext=(10, 5), fontsize=9)

plt.title("Régression Linéaire — Prix des Maisons", fontsize=13, fontweight='bold')
plt.xlabel("Taille (m²)")
plt.ylabel("Prix")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("=" * 50)
print("INTERPRÉTATION")
print("=" * 50)
print(f"Pente     : {slope:.2f}")
print(f"→ Chaque m² supplémentaire augmente le prix de {slope:,.0f}")
print(f"\nOrdonnée  : {intercept:.2f}")
print(f"→ Prix de base théorique (0 m²) = {intercept:,.0f}")
print(f"\nR²        : {r_value**2:.4f}")
print(f"→ Le modèle explique {r_value**2*100:.1f}% de la variance des prix")
print(f"\nPrédiction 90 m² : {prix_predit:,.0f}")
```

```python
#EXERCICE 6

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Données
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# ANOVA
f_stat, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"F-statistique : {f_stat:.4f}")
print(f"P-value       : {p_value:.6f}")

# Interprétation
alpha = 0.05
print(f"\nSeuil alpha   : {alpha}")
if p_value < alpha:
    print("✅ On rejette H0 → les engrais ont des effets significativement différents")
else:
    print("❌ On ne rejette pas H0 → pas de différence significative")
groupes = {
    'Engrais 1': fertilizer_1,
    'Engrais 2': fertilizer_2,
    'Engrais 3': fertilizer_3
}

print("\n=== Statistiques par groupe ===")
for nom, data in groupes.items():
    print(f"{nom} — Moyenne : {np.mean(data):.2f} | Écart-type : {np.std(data):.2f}")
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].boxplot(groupes.values(), labels=groupes.keys(),
                patch_artist=True,
                boxprops=dict(facecolor='steelblue', alpha=0.6))
axes[0].set_title("Distribution par engrais")
axes[0].set_ylabel("Croissance (cm)")
axes[0].grid(True, alpha=0.3)

moyennes = [np.mean(d) for d in groupes.values()]
couleurs  = ['#60A5FA', '#4ADE80', '#F97316']
axes[1].bar(groupes.keys(), moyennes, color=couleurs, edgecolor='white')
axes[1].set_title("Croissance moyenne par engrais")
axes[1].set_ylabel("Moyenne (cm)")
axes[1].axhline(np.mean(moyennes), color='red',
                linestyle='--', label='Moyenne globale')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.suptitle("ANOVA — Effet des engrais sur la croissance", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
print("=" * 50)
print("RÉPONSES AUX QUESTIONS")
print("=" * 50)

print(f"\n1. Valeurs F et p :")
print(f"   F-statistique : {f_stat:.4f}")
print(f"   P-value       : {p_value:.6f}")

print(f"\n2. Effets significatifs ?")
print(f"   p-value ({p_value:.6f}) < 0.05")
print(f"   → OUI, les engrais ont des effets significativement différents")
print(f"   → Engrais 2 est le plus efficace (moyenne = {np.mean(fertilizer_2):.1f} cm)")

print(f"\n3. Si p-value > 0.05 :")
print(f"   → On ne rejetterait pas H0")
print(f"   → On conclurait que les 3 engrais ont des effets similaires")
print(f"   → Les différences observées seraient dues au hasard")
```

```python
#EXERCICE 7

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Paramètres
n = 10       
p = 0.5      
k = 5        

# Probabilité exacte d'obtenir 5 faces
prob_exacte = binom.pmf(k, n, p)
print(f"P(exactement {k} faces en {n} lancers) : {prob_exacte:.4f} ({prob_exacte*100:.2f}%)")

# Probabilités pour tous les cas (0 à 10)
print("\n=== Probabilités pour chaque résultat ===")
for i in range(n + 1):
    prob = binom.pmf(i, n, p)
    barre = "█" * int(prob * 100)
    print(f"  {i:2d} faces : {prob:.4f} ({prob*100:5.2f}%) {barre}")
print("\n=== Probabilités cumulées ===")
print(f"P(X ≤ 5)  : {binom.cdf(5, n, p):.4f} ({binom.cdf(5, n, p)*100:.2f}%)")
print(f"P(X ≥ 5)  : {1 - binom.cdf(4, n, p):.4f} ({(1 - binom.cdf(4, n, p))*100:.2f}%)")
print(f"P(X < 5)  : {binom.cdf(4, n, p):.4f} ({binom.cdf(4, n, p)*100:.2f}%)")
print(f"P(X > 5)  : {1 - binom.cdf(5, n, p):.4f} ({(1 - binom.cdf(5, n, p))*100:.2f}%)")
x    = np.arange(0, n + 1)
probs = binom.pmf(x, n, p)

plt.figure(figsize=(10, 6))
couleurs = ['#F97316' if i == k else '#60A5FA' for i in x]
bars = plt.bar(x, probs, color=couleurs, edgecolor='white', width=0.6)

# Annotation
for bar, prob in zip(bars, probs):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 0.002,
             f'{prob:.3f}', ha='center', va='bottom', fontsize=9)

plt.axhline(prob_exacte, color='red', linestyle='--',
            alpha=0.5, label=f'P(X=5) = {prob_exacte:.4f}')
plt.title(f"Distribution Binomiale (n={n}, p={p})\nOrange = P(X={k})", fontsize=13, fontweight='bold')
plt.xlabel("Nombre de faces")
plt.ylabel("Probabilité")
plt.xticks(x)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
moyenne    = binom.mean(n, p)
variance   = binom.var(n, p)
ecart_type = binom.std(n, p)

print("\n=== Statistiques de la distribution ===")
print(f"Moyenne    : {moyenne:.2f}")
print(f"Variance   : {variance:.2f}")
print(f"Écart-type : {ecart_type:.2f}")
```

```python
#EXERCICE 8

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Données
data = pd.DataFrame({
    'age':    [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

print("Dataset")
print(data)

pearson_r, pearson_p = stats.pearsonr(data['age'], data['income'])

# Corrélation de Spearman
spearman_r, spearman_p = stats.spearmanr(data['age'], data['income'])

print("\n Corrélation de Pearson")
print(f"Coefficient r : {pearson_r:.4f}")
print(f"P-value       : {pearson_p:.6f}")

print("\n=== Corrélation de Spearman ===")
print(f"Coefficient r : {spearman_r:.4f}")
print(f"P-value       : {spearman_p:.6f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter plot avec droite de régression
axes[0].scatter(data['age'], data['income'],
                color='steelblue', s=100, zorder=5)
z = np.polyfit(data['age'], data['income'], 1)
p = np.poly1d(z)
x_line = np.linspace(20, 45, 100)
axes[0].plot(x_line, p(x_line), color='red',
             linewidth=2, linestyle='--')

for _, row in data.iterrows():
    axes[0].annotate(f"({int(row['age'])}, {int(row['income']):,})",
                     (row['age'], row['income']),
                     textcoords="offset points",
                     xytext=(8, 5), fontsize=8)

axes[0].set_title(f"Pearson r = {pearson_r:.4f}\nSpearman r = {spearman_r:.4f}")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Revenu ($)")
axes[0].grid(True, alpha=0.3)

# Heatmap de corrélation
corr_matrix = data.corr()
im = axes[1].imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(im, ax=axes[1])
axes[1].set_xticks([0, 1])
axes[1].set_yticks([0, 1])
axes[1].set_xticklabels(['age', 'income'])
axes[1].set_yticklabels(['age', 'income'])
for i in range(2):
    for j in range(2):
        axes[1].text(j, i, f"{corr_matrix.iloc[i, j]:.4f}",
                     ha='center', va='center',
                     color='white', fontsize=14, fontweight='bold')
axes[1].set_title("Heatmap de corrélation")

plt.suptitle("Corrélations Pearson & Spearman — Age vs Revenu",
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n=== Interprétation ===")
print(f"Pearson  r = {pearson_r:.4f} → corrélation linéaire  {'forte' if abs(pearson_r) > 0.8 else 'modérée'} {'positive' if pearson_r > 0 else 'négative'}")
print(f"Spearman r = {spearman_r:.4f} → corrélation de rang   {'forte' if abs(spearman_r) > 0.8 else 'modérée'} {'positive' if spearman_r > 0 else 'négative'}")
print(f"\nP-value Pearson  : {pearson_p:.6f} → {'significatif ✅' if pearson_p < 0.05 else 'non significatif ❌'}")
print(f"P-value Spearman : {spearman_p:.6f} → {'significatif ✅' if spearman_p < 0.05 else 'non significatif ❌'}")
```
