```python
from functools import wraps

def once(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        if not wrapper.called:
            wrapper.called = True
            return func(*args, **kwargs)
        return None
    wrapper.called = False
    return wrapper
```
