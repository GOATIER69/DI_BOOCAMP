```python
from collections import OrderedDict

# Avec un dictionnaire standard (Python 3.7+ conserve l'ordre)
d = {'nom': 'Dupont', 'age': 30, 'ville': 'Paris'}

# Avec OrderedDict (Python < 3.7)
od = OrderedDict()
od['nom'] = 'Dupont'
od['age'] = 30
od['ville'] = 'Paris'

# Comparaison d'ordre
print(od == OrderedDict([('nom', 'Dupont'), ('age', 30), ('ville', 'Paris')]))  # True
print(od == OrderedDict([('ville', 'Paris'), ('nom', 'Dupont'), ('age', 30)]))  # False
```

```python
# Sans defaultdict
compteur = {}
mots = ['pomme', 'banane', 'pomme', 'orange', 'banane', 'pomme']
for mot in mots:
    if mot in compteur:
        compteur[mot] += 1
    else:
        compteur[mot] = 1
```

```python
from collections import defaultdict

# 1. Comptage d'occurrences
compteur = defaultdict(int)
mots = ['pomme', 'banane', 'pomme', 'orange', 'banane', 'pomme']
for mot in mots:
    compteur[mot] += 1  # Pas besoin de vérifier si la clé existe
print(dict(compteur))  # {'pomme': 3, 'banane': 2, 'orange': 1}

# 2. Regroupement de données
groupes = defaultdict(list)
nombres = [1, 2, 3, 4, 5, 6]
for n in nombres:
    groupes['pair' if n % 2 == 0 else 'impair'].append(n)
print(dict(groupes))  # {'impair': [1, 3, 5], 'pair': [2, 4, 6]}

# 3. Valeur par défaut personnalisée
d = defaultdict(lambda: "clé inexistante")
print(d['inexistant'])  # "clé inexistante"
```
