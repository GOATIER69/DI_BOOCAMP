```python
import random

def number_guessing_game():
    """
    Fonction principale du jeu de devinettes de nombres.
    Génère un nombre aléatoire entre 1 et 100 et laisse l'utilisateur
    essayer de le deviner avec un nombre limité de tentatives.
    """
    # Génération du nombre aléatoire
    random_number = random.randint(1, 100)

    # Définition du nombre maximal de tentatives
    max_attempts = 7

    print("Bienvenue dans le jeu de devinettes de nombres !")
    print("J'ai choisi un nombre entre 1 et 100.")
    print(f"Vous avez {max_attempts} tentatives pour le deviner.\n")

    # Boucle pour les tentatives
    for attempt in range(1, max_attempts + 1):
        print(f"--- Tentative {attempt} sur {max_attempts} ---")

        # Saisie de l'utilisateur
        try:
            guess = int(input("Votre proposition : "))
        except ValueError:
            print("Veuillez entrer un nombre valide.\n")
            continue

        # Comparaison et retour d'information
        if guess < random_number:
            print("Trop bas ! (Too low!)\n")
        elif guess > random_number:
            print("Trop haut ! (Too high!)\n")
        else:
            print(f"Félicitations ! Vous avez deviné le nombre {random_number} en {attempt} tentative(s) !")
            break
    else:
        # Cette clause s'exécute si la boucle se termine sans avoir rencontré 'break'
        print(f"Dommage ! Vous avez utilisé vos {max_attempts} tentatives.")
        print(f"Le nombre mystère était : {random_number}")

# Appel de la fonction pour démarrer le jeu
if __name__ == "__main__":
    number_guessing_game()
```
