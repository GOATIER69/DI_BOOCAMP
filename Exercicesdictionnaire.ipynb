```python
# 1. Créer un dictionnaire à partir d'une liste de tuples
liste_tuples = [("name", "Elie"), ("job", "Instructor")]
dictionnaire1 = {cle: valeur for cle, valeur in liste_tuples}
print(dictionnaire1)

# 2. Créer un dictionnaire à partir de deux listes avec zip()
abreviation = ["CA", "NJ", "RI"]
etats = ["California", "New Jersey", "Rhode Island"]
dictionnaire2 = {abrev: etat for abrev, etat in zip(abreviation, etats)}
print(dictionnaire2)

# 3. Dictionnaire avec les voyelles comme clés et 0 comme valeur
voyelles = "aeiou"
dictionnaire3 = {voyelle: 0 for voyelle in voyelles}
print(dictionnaire3)

# 4. Dictionnaire avec la position et la lettre de l'alphabet
dictionnaire4 = {i: chr(64 + i) for i in range(1, 27)}
print(dictionnaire4)

# Super Bonus : Compter les occurrences des voyelles dans "awesome sauce"
chaine = "awesome sauce"
voyelles_bonus = "aeiou"
dictionnaire_bonus = {voyelle: sum(1 for lettre in chaine.lower() if lettre == voyelle)
                      for voyelle in voyelles_bonus}
print(dictionnaire_bonus)
```
