```python
from collections import deque
import heapq
from dataclasses import dataclass, field
from typing import Any
import time

@dataclass(order=True)
class Task:
    """Classe représentant une tâche avec sa priorité et sa description"""
    priority: int
    description: str = field(compare=False)
    timestamp: float = field(default_factory=time.time, compare=False)
    id: int = field(default_factory=lambda: id(object()), compare=False)

class TaskScheduler:
    """Planificateur de tâches avec file d'attente prioritaire"""

    def __init__(self):
        # Utilisation d'un tas (heap) pour la gestion des priorités
        self.task_heap = []
        self.counter = 0

    def add_task(self, description: str, priority: int) -> None:
        """
        Ajoute une tâche à la file d'attente

        Args:
            description: Description de la tâche
            priority: Niveau de priorité (plus petit = plus prioritaire)
        """
        task = Task(priority=priority, description=description)
        # On utilise un compteur pour gérer l'ordre FIFO pour les priorités égales
        heapq.heappush(self.task_heap, (priority, self.counter, task))
        self.counter += 1
        print(f"✅ Tâche ajoutée : '{description}' (priorité {priority})")

    def execute_next_task(self) -> str:
        """
        Exécute la tâche suivante en fonction de sa priorité

        Returns:
            str: Description de la tâche exécutée
        """
        if not self.task_heap:
            print("❌ Aucune tâche à exécuter")
            return None

        priority, counter, task = heapq.heappop(self.task_heap)
        print(f"🚀 Exécution de la tâche : '{task.description}' (priorité {priority})")
        return task.description

    def view_queue(self) -> None:
        """Affiche la file d'attente des tâches en cours"""
        if not self.task_heap:
            print("📭 La file d'attente est vide")
            return

        print("\n📋 File d'attente des tâches :")
        print("-" * 50)
        # Tri par priorité puis par ordre d'arrivée
        sorted_tasks = sorted(self.task_heap, key=lambda x: (x[0], x[1]))
        for priority, counter, task in sorted_tasks:
            print(f"   - {task.description} (priorité {priority})")
        print("-" * 50)
        print(f"Total : {len(self.task_heap)} tâche(s) en attente\n")

    def get_queue_size(self) -> int:
        """Retourne le nombre de tâches en attente"""
        return len(self.task_heap)

    def clear_queue(self) -> None:
        """Vide la file d'attente"""
        self.task_heap.clear()
        self.counter = 0
        print("🗑️ File d'attente vidée")

# Exemple d'utilisation et démonstration
def demo_scheduler():
    """Démonstration du planificateur de tâches"""

    print("=" * 60)
    print("📅 PLANIFICATEUR DE TÂCHES - DÉMONSTRATION")
    print("=" * 60)

    # Création du planificateur
    scheduler = TaskScheduler()

    # Ajout de tâches avec différentes priorités
    print("\n📥 AJOUT DE TÂCHES :")
    scheduler.add_task("Envoyer l'email de bienvenue", priority=2)
    scheduler.add_task("Sauvegarder la base de données", priority=1)  # Priorité élevée
    scheduler.add_task("Faire le ménage", priority=3)
    scheduler.add_task("Répondre aux emails urgents", priority=1)  # Priorité élevée
    scheduler.add_task("Mettre à jour le site web", priority=2)

    # Affichage de la file d'attente
    scheduler.view_queue()

    # Exécution des tâches une par une
    print("🔄 EXÉCUTION DES TÂCHES :")
    for i in range(6):  # On essaie d'exécuter 6 tâches (il n'y en a que 5)
        scheduler.execute_next_task()
        print("Tâches restantes :", scheduler.get_queue_size())

    # Nouveau cycle de tâches
    print("\n📥 AJOUT DE NOUVELLES TÂCHES :")
    scheduler.add_task("Compte-rendu mensuel", priority=2)
    scheduler.add_task("Mise à jour antivirus", priority=0)  # Priorité maximale
    scheduler.add_task("Vérification des sauvegardes", priority=1)

    scheduler.view_queue()

    # Exécution prioritaire
    print("🔄 EXÉCUTION PRIORITAIRE :")
    scheduler.execute_next_task()  # Devrait être "Mise à jour antivirus"
    scheduler.execute_next_task()  # Devrait être "Vérification des sauvegardes"

    scheduler.view_queue()

if __name__ == "__main__":
    # Exécution de la démonstration
    demo_scheduler()

    print("\n" + "=" * 60)
    print("✨ FIN DE LA DÉMONSTRATION")
    print("=" * 60)
```

```python
from collections import deque
import re

def check_palindrome(text: str) -> bool:
    """
    Vérifie si un mot ou une phrase est un palindrome en utilisant une file d'attente.

    Args:
        text: Le texte à vérifier

    Returns:
        bool: True si c'est un palindrome, False sinon
    """
    # 1. Nettoyer le texte : ignorer les espaces, ponctuations et la casse
    cleaned_text = re.sub(r'[^a-zA-Z0-9]', '', text).lower()

    # Si le texte est vide ou un caractère, c'est un palindrome
    if len(cleaned_text) <= 1:
        return True

    # 2. Créer une file d'attente avec tous les caractères
    char_queue = deque(cleaned_text)

    # 3. Vérifier si la file est un palindrome
    while len(char_queue) > 1:
        # Retirer le premier caractère (gauche)
        left = char_queue.popleft()
        # Retirer le dernier caractère (droite)
        right = char_queue.pop()

        # Si les caractères ne correspondent pas, ce n'est pas un palindrome
        if left != right:
            return False

    # Si on a parcouru toute la file, c'est un palindrome
    return True

# Version alternative avec explanation détaillée
def check_palindrome_detailed(text: str) -> tuple[bool, str]:
    """
    Version détaillée qui montre le processus étape par étape.

    Returns:
        tuple: (est_un_palindrome, message_explicatif)
    """
    # Nettoyage du texte
    cleaned = re.sub(r'[^a-zA-Z0-9]', '', text).lower()
    original_display = f"'{text}'" if text != cleaned else f"'{text}'"

    if len(cleaned) <= 1:
        return True, f"{original_display} est un palindrome (1 caractère ou vide)"

    # Création de la file
    q = deque(cleaned)
    steps = []
    steps.append(f"Texte nettoyé : '{cleaned}'")

    # Comparaison
    is_palindrome = True
    while len(q) > 1:
        left = q.popleft()
        right = q.pop()
        steps.append(f"Comparaison : '{left}' vs '{right}'")

        if left != right:
            steps.append(f"❌ Différence détectée : '{left}' ≠ '{right}'")
            is_palindrome = False
            break

    if is_palindrome:
        steps.append(f"✅ Toutes les paires correspondent !")
        return True, f"{original_display} est un palindrome !\n" + "\n".join(steps)
    else:
        return False, f"{original_display} n'est pas un palindrome.\n" + "\n".join(steps)

# Tests et démonstration
def test_palindrome_checker():
    """Teste la fonction avec différents exemples"""

    test_cases = [
        ("Racecar", True),
        ("hello", False),
        ("A man a plan a canal Panama", True),
        ("Never odd or even", True),
        ("Madam", True),
        ("No lemon, no melon", True),
        ("Was it a car or a cat I saw?", True),
        ("Python", False),
        ("", True),
        ("a", True),
        ("ab", False),
        ("Able was I ere I saw Elba", True),
        ("12321", True),
        ("12345", False),
    ]

    print("=" * 70)
    print("🧪 TEST DU VÉRIFICATEUR DE PALINDROMES")
    print("=" * 70)

    # Test standard
    print("\n📊 TESTS RAPIDES :")
    for text, expected in test_cases:
        result = check_palindrome(text)
        status = "✅" if result == expected else "❌"
        print(f"{status} check_palindrome({text!r:35}) -> {result} (attendu: {expected})")

    # Tests détaillés
    print("\n" + "=" * 70)
    print("🔍 ANALYSE DÉTAILLÉE DE QUELQUES CAS :")
    print("=" * 70)

    detailed_cases = [
        "Racecar",
        "hello",
        "A man a plan a canal Panama",
        "No lemon, no melon",
        "Was it a car or a cat I saw?"
    ]

    for text in detailed_cases:
        print(f"\n📝 Test : {text!r}")
        print("-" * 50)
        result, explanation = check_palindrome_detailed(text)
        print(explanation)
        print(f"Résultat final : {result}")

# Fonction interactive
def interactive_palindrome_check():
    """Mode interactif pour tester des phrases personnalisées"""

    print("\n" + "=" * 70)
    print("🎮 MODE INTERACTIF - VÉRIFICATEUR DE PALINDROMES")
    print("=" * 70)
    print("Entrez 'quit' pour quitter")
    print("Entrez 'demo' pour voir des exemples")
    print("-" * 70)

    while True:
        user_input = input("\n📝 Entrez un mot ou une phrase : ").strip()

        if user_input.lower() == 'quit':
            print("👋 Au revoir !")
            break
        elif user_input.lower() == 'demo':
            test_palindrome_checker()
            continue
        elif not user_input:
            print("⚠️ Veuillez entrer un texte valide.")
            continue

        # Analyse détaillée
        is_pal, explanation = check_palindrome_detailed(user_input)
        print("\n" + "-" * 50)
        print(explanation)
        print("-" * 50)

        if is_pal:
            print("🎉 C'EST UN PALINDROME !")
        else:
            print("❌ Ce n'est PAS un palindrome.")

# Implémentation avec file d'attente uniquement (sans regex, plus pédagogique)
def check_palindrome_simple(text: str) -> bool:
    """
    Version simple sans regex, utilisant uniquement les méthodes de base.
    """
    # Nettoyage manuel
    cleaned = ''.join(char.lower() for char in text if char.isalnum())

    if len(cleaned) <= 1:
        return True

    q = deque(cleaned)

    while len(q) > 1:
        if q.popleft() != q.pop():
            return False

    return True

if __name__ == "__main__":
    # Exécution des tests
    test_palindrome_checker()

    # Démonstration de la version simple
    print("\n" + "=" * 70)
    print("📝 VERSION SIMPLE (sans regex) :")
    print("=" * 70)
    print(f"check_palindrome_simple('Racecar') -> {check_palindrome_simple('Racecar')}")
    print(f"check_palindrome_simple('hello') -> {check_palindrome_simple('hello')}")
    print(f"check_palindrome_simple('A man a plan a canal Panama') -> {check_palindrome_simple('A man a plan a canal Panama')}")

    # Mode interactif (décommentez pour utiliser)
    # interactive_palindrome_check()
```

```python
class Node:
    """
    Classe représentant un nœud dans une liste chaînée.
    """
    def __init__(self, value):
        self.value = value
        self.next = None


class LinkedList:
    """
    Classe représentant une liste chaînée avec des opérations de base.
    """
    def __init__(self):
        self.head = None

    def append(self, value):
        """
        Ajoute un nouveau nœud à la fin de la liste.

        Args:
            value: La valeur à ajouter
        """
        new_node = Node(value)

        # Si la liste est vide, le nouveau nœud devient la tête
        if self.head is None:
            self.head = new_node
            return

        # Parcourir la liste jusqu'au dernier nœud
        current = self.head
        while current.next is not None:
            current = current.next

        # Attacher le nouveau nœud à la fin
        current.next = new_node

    def to_list(self):
        """
        Convertit la liste chaînée en une liste Python pour l'affichage.

        Returns:
            list: Liste Python contenant toutes les valeurs
        """
        result = []
        current = self.head
        while current is not None:
            result.append(current.value)
            current = current.next
        return result

    def display(self):
        """
        Affiche la liste chaînée de manière lisible.
        """
        values = self.to_list()
        if not values:
            print("Liste vide")
        else:
            print(" -> ".join(str(v) for v in values))


def merge_sorted_linked_lists(list1, list2):
    """
    Fusionne deux listes chaînées triées en une seule liste triée.

    Args:
        list1 (LinkedList): Première liste chaînée triée
        list2 (LinkedList): Deuxième liste chaînée triée

    Returns:
        Node: Tête de la nouvelle liste chaînée triée

    Time Complexity: O(n + m) où n et m sont les tailles des listes
    Space Complexity: O(1) (on réutilise les nœuds existants)
    """
    # Cas particuliers : une ou les deux listes sont vides
    if list1.head is None:
        return list2.head
    if list2.head is None:
        return list1.head

    # Nœud factice (dummy) pour faciliter la construction
    dummy = Node(0)
    current = dummy

    # Pointeurs pour parcourir les deux listes
    p1 = list1.head
    p2 = list2.head

    # Parcourir les deux listes simultanément
    while p1 is not None and p2 is not None:
        if p1.value <= p2.value:
            current.next = p1
            p1 = p1.next
        else:
            current.next = p2
            p2 = p2.next
        current = current.next

    # Si une liste est épuisée, ajouter les nœuds restants de l'autre
    if p1 is not None:
        current.next = p1
    elif p2 is not None:
        current.next = p2

    # Retourner la tête de la nouvelle liste (sans le nœud factice)
    return dummy.next


def merge_sorted_linked_lists_recursive(list1, list2):
    """
    Version récursive de la fusion de deux listes chaînées triées.

    Args:
        list1 (LinkedList): Première liste chaînée triée
        list2 (LinkedList): Deuxième liste chaînée triée

    Returns:
        Node: Tête de la nouvelle liste chaînée triée
    """
    # Cas de base
    if list1.head is None:
        return list2.head
    if list2.head is None:
        return list1.head

    # Fusion récursive
    def merge_nodes(node1, node2):
        if node1 is None:
            return node2
        if node2 is None:
            return node1

        if node1.value <= node2.value:
            node1.next = merge_nodes(node1.next, node2)
            return node1
        else:
            node2.next = merge_nodes(node1, node2.next)
            return node2

    return merge_nodes(list1.head, list2.head)


def create_linked_list_from_list(values):
    """
    Fonction utilitaire pour créer une LinkedList à partir d'une liste Python.

    Args:
        values (list): Liste de valeurs

    Returns:
        LinkedList: Liste chaînée créée
    """
    ll = LinkedList()
    for val in values:
        ll.append(val)
    return ll


def test_merge_function():
    """
    Fonction de test complète pour valider la fusion des listes chaînées.
    """
    print("=" * 70)
    print("🔗 FUSION DE LISTES CHAÎNÉES TRIÉES")
    print("=" * 70)

    # Test 1: Cas normal
    print("\n📝 Test 1: Fusion de deux listes de même taille")
    print("-" * 50)
    list1 = create_linked_list_from_list([1, 3, 5, 7])
    list2 = create_linked_list_from_list([2, 4, 6, 8])

    print("Liste 1:", end=" ")
    list1.display()
    print("Liste 2:", end=" ")
    list2.display()

    merged_head = merge_sorted_linked_lists(list1, list2)
    merged_list = LinkedList()
    merged_list.head = merged_head
    print("Liste fusionnée:", end=" ")
    merged_list.display()

    # Test 2: Listes de tailles différentes
    print("\n📝 Test 2: Listes de tailles différentes")
    print("-" * 50)
    list1 = create_linked_list_from_list([1, 2, 3])
    list2 = create_linked_list_from_list([4, 5, 6, 7, 8])

    print("Liste 1:", end=" ")
    list1.display()
    print("Liste 2:", end=" ")
    list2.display()

    merged_head = merge_sorted_linked_lists(list1, list2)
    merged_list = LinkedList()
    merged_list.head = merged_head
    print("Liste fusionnée:", end=" ")
    merged_list.display()

    # Test 3: Une liste vide
    print("\n📝 Test 3: Une liste vide")
    print("-" * 50)
    list1 = create_linked_list_from_list([])
    list2 = create_linked_list_from_list([1, 2, 3, 4, 5])

    print("Liste 1:", end=" ")
    list1.display()
    print("Liste 2:", end=" ")
    list2.display()

    merged_head = merge_sorted_linked_lists(list1, list2)
    merged_list = LinkedList()
    merged_list.head = merged_head
    print("Liste fusionnée:", end=" ")
    merged_list.display()

    # Test 4: Liste avec des doublons
    print("\n📝 Test 4: Liste avec des valeurs dupliquées")
    print("-" * 50)
    list1 = create_linked_list_from_list([1, 3, 3, 5])
    list2 = create_linked_list_from_list([2, 3, 4, 6])

    print("Liste 1:", end=" ")
    list1.display()
    print("Liste 2:", end=" ")
    list2.display()

    merged_head = merge_sorted_linked_lists(list1, list2)
    merged_list = LinkedList()
    merged_list.head = merged_head
    print("Liste fusionnée:", end=" ")
    merged_list.display()

    # Test 5: Version récursive
    print("\n📝 Test 5: Version récursive")
    print("-" * 50)
    list1 = create_linked_list_from_list([1, 3, 5])
    list2 = create_linked_list_from_list([2, 4, 6])

    print("Liste 1:", end=" ")
    list1.display()
    print("Liste 2:", end=" ")
    list2.display()

    merged_head_rec = merge_sorted_linked_lists_recursive(list1, list2)
    merged_list_rec = LinkedList()
    merged_list_rec.head = merged_head_rec
    print("Liste fusionnée (récursive):", end=" ")
    merged_list_rec.display()


def interactive_merge():
    """
    Mode interactif pour fusionner des listes personnalisées.
    """
    print("\n" + "=" * 70)
    print("🎮 MODE INTERACTIF - FUSION DE LISTES CHAÎNÉES")
    print("=" * 70)
    print("Entrez des nombres séparés par des virgules")
    print("Exemple: 1, 3, 5, 7")
    print("-" * 70)

    try:
        # Saisie de la première liste
        input1 = input("\n📥 Entrez la première liste triée: ").strip()
        if input1:
            values1 = [int(x.strip()) for x in input1.split(',')]
        else:
            values1 = []

        # Saisie de la deuxième liste
        input2 = input("📥 Entrez la deuxième liste triée: ").strip()
        if input2:
            values2 = [int(x.strip()) for x in input2.split(',')]
        else:
            values2 = []

        # Création des listes chaînées
        list1 = create_linked_list_from_list(values1)
        list2 = create_linked_list_from_list(values2)

        print("\n📊 Résultat:")
        print("-" * 50)
        print("Liste 1:", end=" ")
        list1.display()
        print("Liste 2:", end=" ")
        list2.display()

        # Fusion
        merged_head = merge_sorted_linked_lists(list1, list2)
        merged_list = LinkedList()
        merged_list.head = merged_head

        print("Liste fusionnée:", end=" ")
        merged_list.display()

    except ValueError:
        print("❌ Erreur: Veuillez entrer des nombres valides séparés par des virgules.")


if __name__ == "__main__":
    # Exécution des tests
    test_merge_function()

    # Exemple de base comme demandé dans l'énoncé
    print("\n" + "=" * 70)
    print("📘 EXEMPLE DE L'ÉNONCÉ")
    print("=" * 70)

    # List1: 1 -> 3 -> 5
    # List2: 2 -> 4 -> 6
    list1 = create_linked_list_from_list([1, 3, 5])
    list2 = create_linked_list_from_list([2, 4, 6])

    print("Liste 1:", end=" ")
    list1.display()
    print("Liste 2:", end=" ")
    list2.display()

    merged_head = merge_sorted_linked_lists(list1, list2)
    merged_list = LinkedList()
    merged_list.head = merged_head

    print("Liste fusionnée:", end=" ")
    merged_list.display()

    # Mode interactif (décommentez pour utiliser)
    # interactive_merge()
```
