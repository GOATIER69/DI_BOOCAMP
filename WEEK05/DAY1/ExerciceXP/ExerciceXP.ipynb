EXERCICE 1

1. Principale différence entre l'apprentissage automatique traditionnel et l'apprentissage profond
L'apprentissage automatique traditionnel nécessite une ingénierie des caractéristiques manuelle, où un expert humain extrait les features pertinentes. L'apprentissage profond apprend automatiquement les représentations hiérarchiques des caractéristiques directement à partir des données brutes, sans intervention manuelle.

2. Comment les RNA imitent-ils le cerveau humain
Les RNA imitent le cerveau par des neurones artificiels interconnectés qui reçoivent des signaux d'entrée, les traitent via des fonctions d'activation, et transmettent des sorties à d'autres neurones. L'apprentissage par ajustement des poids synaptiques (connexions) correspond au renforcement ou affaiblissement des connexions synaptiques biologiques.

3. Pourquoi l'apprentissage profond est plus performant sur les grands ensembles de données
L'apprentissage profond possède une très grande capacité de paramétrage (millions voire milliards de paramètres). Sur de petits jeux de données, il surcharge. Sur de grands jeux de données, il généralise bien car la régularisation implicite des réseaux profonds permet d'exploiter la structure statistique des données massives. Les modèles traditionnels plafonnent plus rapidement.

4. Défis de l'apprentissage profond et solutions

Surapprentissage : régularisation (dropout, L1/L2), data augmentation, early stopping

Gradient vanishing/exploding : normalisation par lots (batch norm), fonctions d'activation ReLU, initialisation adaptée

Coût computationnel : GPUs/TPUs, calcul distribué

Données massives nécessaires : transfert learning, data augmentation, synthèse de données

Interprétabilité : techniques d'explicabilité (SHAP, LIME, attention maps)

5. Ingénierie des caractéristiques et pourquoi non nécessaire en deep learning
L'ingénierie des caractéristiques est le processus de transformation manuelle des données brutes en features pertinentes pour un modèle. L'apprentissage profond l'évite car les couches cachées apprennent automatiquement des représentations hiérarchiques : les premières couches détectent des patterns simples (bords, textures), les couches profondes combinent ces patterns en concepts abstraits.

6. Rôle des couches cachées
Les couches cachées transforment progressivement l'entrée en représentations de plus en plus abstraites. Chaque couche applique une transformation non-linéaire. L'empilement des couches permet d'apprendre des fonctions complexes hiérarchiques. Plus de couches (profondeur) permet de représenter des relations plus sophistiquées avec moins de neurones par couche.

7. Fonction d'une fonction d'activation
Une fonction d'activation introduit la non-linéarité dans le réseau. Sans elle, un réseau multicouche serait équivalent à une simple régression linéaire (composition de fonctions linéaires). Elle permet au réseau d'apprendre des relations non-linéaires complexes. Exemples courants : ReLU, sigmoïde, tanh.

EXERCICE 2

Cas 1 : Température = 70°F, Pluie = 0
Somme pondérée = (70 × 0,6) + (0 × 0,4) + 2 = 42 + 0 + 2 = 44
44 > 20 → Sortie = 1 (Oui, sortir)

Cas 2 : Température = 50°F, Pluie = 1
Somme pondérée = (50 × 0,6) + (1 × 0,4) + 2 = 30 + 0,4 + 2 = 32,4
32,4 > 20 → Sortie = 1 (Oui, sortir)

Interprétation :
Le perceptron a suggéré de sortir dans les deux cas. La raison est que le poids de la température (0,6) est suffisamment élevé et le biais (2) est positif. Même avec une température de 50°F et une pluie présente, la somme pondérée (32,4) dépasse largement le seuil de 20. Ce modèle n'est pas réaliste car il ne pénalise pas suffisamment la pluie. Pour un meilleur modèle, il faudrait augmenter le poids de la pluie négativement ou augmenter le seuil.

```python
#EXERCICE 3

# 1. Importation des modules
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# 2. Chargement du jeu de données MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 3. Normalisation des données
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# 4. Encodage one-hot des étiquettes
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 5. Construction du modèle séquentiel
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 6. Compilation du modèle
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 7. Entraînement du modèle
history = model.fit(
    x_train, y_train,
    batch_size=32,
    epochs=5,
    validation_split=0.1,
    verbose=1
)

# Évaluation sur l'ensemble de test
test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Précision sur l'ensemble de test : {test_accuracy:.4f}")
```

La superficie contribue à hauteur de 1000 (2000 × 0,5) et le nombre de chambres contribue à hauteur de 2,1. Le biais de 50 000 représente le prix de base. Ce modèle donne une contribution très faible aux chambres par rapport à la superficie.

```python
#EXERCICE 5

import numpy as np

# Initialize input data (features)
x = np.array([4, 80])  # 4 hours studied, previous test score: 80

# Initialize weights and bias
w = np.array([0.6, 0.3])  # Initial weights
b = 10  # Initial bias

# Forward Propagation
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b  # Weighted sum
    return z  # Linear activation (No ReLU here, it's a regression task)

# Compute prediction
y_pred = forward_propagation(x, w, b)
y_true = 85  # Actual exam score

# Compute Loss (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

# Compute Gradients
grad_w = -(y_true - y_pred) * x  # Partial derivatives with respect to weights
grad_b = -(y_true - y_pred)  # Partial derivative with respect to bias

# Update Weights and Bias
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

# Print Results
print("Initial Prediction:", y_pred)
print("Loss:", loss)
print("Updated Weights:", w_new)
print("Updated Bias:", b_new)
```

Exécution du code :

Updated Bias: 10 - 0.01 * (-48.6) = 10 - (-0.486) = 10.486

Pourquoi la descente de gradient réduit l'erreur :
La descente de gradient calcule la direction de la plus forte augmentation de l'erreur (gradient). En soustrayant le gradient multiplié par le taux d'apprentissage, on se déplace dans la direction opposée, soit celle qui diminue l'erreur. Les gradients indiquent comment modifier chaque paramètre pour réduire la perte.

Impact des modifications :

Taux d'apprentissage plus petit (ex: 0.001) : Apprentissage plus lent, convergence plus longue mais plus stable.

Taux d'apprentissage plus grand (ex: 0.1) : Apprentissage plus rapide mais risque de diverger si trop grand.

Poids initiaux différents : Peut converger vers un minimum différent ou ralentir la convergence.

Poids initiaux nuls : Tous les neurones apprennent la même chose (symétrie), ce qui est problématique.

```python
#EXERCICE 6

# 0. Importation des bibliotheques requises
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# 1. Chargement du jeu de donnees MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Normalisation des donnees
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# 3. Encodage one-hot des etiquettes
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# 4. Construction du modele
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# 5. Compilation du modele
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 6. Entrainement du modele
model.fit(x_train, y_train_cat, batch_size=32, epochs=5, validation_split=0.1, verbose=1)

# 7. Predictions sur les donnees de test
y_pred_proba = model.predict(x_test)
y_pred = np.argmax(y_pred_proba, axis=1)

# 8. Visualisation des predictions
n_images = 10
plt.figure(figsize=(12, 4))
for i in range(n_images):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_test[i], cmap='gray')
    plt.title(f'Pred: {y_pred[i]}\nTrue: {y_test[i]}')
    plt.axis('off')
plt.tight_layout()
plt.show()

# Affichage de la precision sur l'ensemble de test
test_loss, test_acc = model.evaluate(x_test, y_test_cat, verbose=0)
print(f"Precision sur l'ensemble de test: {test_acc:.4f}")
```
