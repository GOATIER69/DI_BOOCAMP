 Tableau comparatif
Aspect	Apprentissage automatique traditionnel	Apprentissage profond
Ingénierie des fonctionnalités	Manuelle et approfondie (feature engineering)	Automatique (extraction hiérarchique des caractéristiques)
Interprétabilité	Élevée (arbres de décision, régression logistique, etc.)	Faible (boîte noire)
Évolutilité	Moins performante sur très grands volumes de données	Très performante avec de grandes quantités de données
Découverte de modèles	Modèles simples (linéaires, SVM, forêts aléatoires)	Modèles complexes (réseaux multi-couches, CNN, RNN, transformers)
Exigences de calcul	Modestes (CPU suffisant)	Élevées (GPU/TPU souvent nécessaires)

2. Problèmes concrets adaptés
Apprentissage automatique traditionnel :
Prédiction du risque de crédit bancaire avec un petit ensemble de données tabulaires contenant des variables bien définies (âge, revenu, historique de paiement). Les modèles comme la régression logistique ou les forêts aléatoires sont interprétables et suffisants.

Apprentissage profond :
Reconnaissance d’objets dans des images satellites (forêts, routes, bâtiments). La variabilité des formes, des couleurs et des éclairages nécessite une extraction automatique des caractéristiques, que l’apprentissage profond réalise efficacement via des couches de convolution.

3. Avantage de l’apprentissage profond pour les données non structurées
L’apprentissage profond excelle sur les données non structurées (images, audio, texte) car il élimine le besoin d’une ingénierie des caractéristiques manuelle, souvent complexe et incomplète. Ses couches hiérarchiques apprennent automatiquement des représentations allant des motifs simples (bords, tons) aux concepts abstraits (visages, phrases). Cette capacité d’extraction end-to-end permet d’obtenir des performances supérieures lorsque de grandes quantités de données sont disponibles.

```python
#EXERCICE 3

# 1. Importation des bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# 2. Génération de l'ensemble de données bruitées
np.random.seed(0)  # Pour la reproductibilité des résultats
x = np.arange(-1, 1, 0.1)  # 20 points de -1 à 1 (exclu)
y = -x**2 + np.random.normal(0, 0.05, len(x))

# Affichage des premières valeurs pour vérification
print("Premières valeurs de x :", x[:5])
print("Premières valeurs de y :", y[:5])
print(f"Nombre total de points : {len(x)}")

# 3. Visualisation des données sur un nuage de points
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.7, label='Données bruitées')
plt.plot(x, -x**2, color='red', linestyle='--', alpha=0.5, label='Fonction réelle sans bruit : y = -x²')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ensemble de données bruitées généré à partir de y = -x²')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 4. Division des données en ensemble d'entraînement et de test
x_train = x[:12]   # Les 12 premiers points
y_train = y[:12]
x_test = x[12:]    # Les 8 derniers points
y_test = y[12:]

print(f"\nTaille de l'ensemble d'entraînement : {len(x_train)} points")
print(f"Taille de l'ensemble de test : {len(x_test)} points")
```

```python
#EXERCICE 4

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Données générées dans l'exercice 3 (à exécuter avant)
np.random.seed(0)
x = np.arange(-1, 1, 0.1)
y = -x**2 + np.random.normal(0, 0.05, len(x))

x_train = x[:12]
y_train = y[:12]
x_test = x[12:]
y_test = y[12:]

# 1. Fonction polynomial_fit
def polynomial_fit(degree):
    """
    Ajuste un polynôme de degré donné aux données d'entraînement.
    
    Paramètres:
    degree : int - degré du polynôme
    
    Retourne:
    polynomial : modèle LinearRegression entraîné avec les features polynomiales
    """
    # Création des features polynomiales
    poly_features = PolynomialFeatures(degree=degree)
    X_train_poly = poly_features.fit_transform(x_train.reshape(-1, 1))
    
    # Ajustement du modèle de régression linéaire
    model = LinearRegression()
    model.fit(X_train_poly, y_train)
    
    return model, poly_features

# 2. Fonction plot_polyfit
def plot_polyfit(degree):
    """
    Trace les données d'entraînement, de test et la courbe polynomiale ajustée.
    
    Paramètres:
    degree : int - degré du polynôme
    """
    # Ajustement du modèle
    model, poly_features = polynomial_fit(degree)
    
    # Prédictions sur l'entraînement et le test
    X_train_poly = poly_features.transform(x_train.reshape(-1, 1))
    X_test_poly = poly_features.transform(x_test.reshape(-1, 1))
    
    y_train_pred = model.predict(X_train_poly)
    y_test_pred = model.predict(X_test_poly)
    
    # Calcul des erreurs
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    
    # Création d'une courbe continue pour la visualisation
    x_smooth = np.linspace(-1, 1, 200)
    X_smooth_poly = poly_features.transform(x_smooth.reshape(-1, 1))
    y_smooth = model.predict(X_smooth_poly)
    
    # Tracé
    plt.figure(figsize=(10, 6))
    plt.scatter(x_train, y_train, color='blue', alpha=0.7, label='Entraînement (12 points)')
    plt.scatter(x_test, y_test, color='green', alpha=0.7, label='Test (8 points)')
    plt.plot(x_smooth, y_smooth, color='red', linewidth=2, label=f'Polynôme degré {degree}')
    plt.plot(x_smooth, -x_smooth**2, color='gray', linestyle='--', alpha=0.5, label='Fonction réelle : y = -x²')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Ajustement polynomial - Degré {degree}\nTrain MSE = {train_mse:.4f} | Test MSE = {test_mse:.4f}')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
    
    return train_mse, test_mse

# 3. Visualisation pour les degrés 1, 7 et 11
print("=== Analyse des ajustements polynomiaux ===\n")

print("1. Degré 1 (sous-apprentissage - linéaire)")
mse_deg1 = plot_polyfit(1)

print("\n2. Degré 7 (bon ajustement)")
mse_deg7 = plot_polyfit(7)

print("\n3. Degré 11 (surapprentissage)")
mse_deg11 = plot_polyfit(11)

# 4. Tableau récapitulatif des observations
print("\n=== Tableau récapitulatif des observations ===")
print("-" * 60)
print(f"{'Degré':<10} {'Train MSE':<15} {'Test MSE':<15} {'Observation'}")
print("-" * 60)
print(f"{1:<10} {mse_deg1[0]:<15.4f} {mse_deg1[1]:<15.4f} Sous-apprentissage : modèle trop simple")
print(f"{7:<10} {mse_deg7[0]:<15.4f} {mse_deg7[1]:<15.4f} Bon ajustement : généralise correctement")
print(f"{11:<10} {mse_deg11[0]:<15.4f} {mse_deg11[1]:<15.4f} Surapprentissage : trop flexible")
print("-" * 60)
```

```python
#EXERCICE 5

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Données générées dans l'exercice 3
np.random.seed(0)
x = np.arange(-1, 1, 0.1)
y = -x**2 + np.random.normal(0, 0.05, len(x))

x_train = x[:12]
y_train = y[:12]
x_test = x[12:]
y_test = y[12:]

# 1 et 2. Parcourir les degrés 1 à 11 et calculer les RMSE
degrees = range(1, 12)
results = []  # Liste pour stocker (degré, train_rmse, test_rmse)

print("=== Calcul des erreurs RMSE pour degrés 1 à 11 ===\n")
print(f"{'Degré':<10} {'Train RMSE':<15} {'Test RMSE':<15}")
print("-" * 40)

for degree in degrees:
    # Création des features polynomiales
    poly_features = PolynomialFeatures(degree=degree)
    X_train_poly = poly_features.fit_transform(x_train.reshape(-1, 1))
    X_test_poly = poly_features.transform(x_test.reshape(-1, 1))
    
    # Ajustement du modèle
    model = LinearRegression()
    model.fit(X_train_poly, y_train)
    
    # Prédictions
    y_train_pred = model.predict(X_train_poly)
    y_test_pred = model.predict(X_test_poly)
    
    # Calcul des RMSE (racine de l'erreur quadratique moyenne)
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    
    results.append((degree, train_rmse, test_rmse))
    print(f"{degree:<10} {train_rmse:<15.6f} {test_rmse:<15.6f}")

# 3. Les résultats sont déjà stockés dans la liste 'results'

# 4. Tracer les courbes RMSE en fonction du degré (échelle logarithmique)
plt.figure(figsize=(10, 6))

degrees_plot = [r[0] for r in results]
train_rmse_vals = [r[1] for r in results]
test_rmse_vals = [r[2] for r in results]

plt.plot(degrees_plot, train_rmse_vals, 'o-', color='blue', label='Entraînement (RMSE)')
plt.plot(degrees_plot, test_rmse_vals, 's-', color='green', label='Test (RMSE)')
plt.xlabel('Degré du polynôme')
plt.ylabel('RMSE (échelle logarithmique)')
plt.yscale('log')  # Échelle logarithmique pour l'axe des ordonnées
plt.title('Évolution de la RMSE en fonction du degré polynomial')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(degrees_plot)
plt.show()

# 5. Identifier le degré optimal (qui minimise la RMSE sur l'ensemble de test)
optimal_degree = min(results, key=lambda x: x[2])[0]
min_test_rmse = min(results, key=lambda x: x[2])[2]

print("\n=== Résultats de la validation croisée ===")
print(f"Degré optimal (minimisant la RMSE de test) : {optimal_degree}")
print(f"RMSE de test minimale : {min_test_rmse:.6f}")
print(f"RMSE d'entraînement correspondante : {results[optimal_degree-1][1]:.6f}")

# Vérification par rapport au modèle sous-jacent
print("\n=== Vérification ===")
print(f"Modèle sous-jacent réel : y = -x² (degré 2)")
if optimal_degree == 2:
    print("✓ Le degré optimal correspond exactement au modèle sous-jacent (degré 2).")
elif abs(optimal_degree - 2) <= 1:
    print(f"✓ Le degré optimal ({optimal_degree}) est proche du modèle sous-jacent (degré 2).")
    print("  Le bruit peut légèrement influencer le choix du degré.")
else:
    print(f"⚠ Le degré optimal ({optimal_degree}) diffère du modèle sous-jacent (degré 2).")
    print("  Vérifiez le bruit ou la taille de l'ensemble de données.")

# Analyse supplémentaire : afficher les valeurs précises pour le degré 2
deg2_results = results[1]  # degré 2 (index 1)
print(f"\n=== Résultats pour le degré 2 (modèle réel) ===")
print(f"Train RMSE : {deg2_results[1]:.6f}")
print(f"Test RMSE  : {deg2_results[2]:.6f}")
```
