```python
# 1. Afficher toutes les valeurs de la liste [1,2,3,4]
liste1 = [1, 2, 3, 4]
print([x for x in liste1])

# 2. Afficher toutes les valeurs multipliées par 20
print([x * 20 for x in liste1])

# 3. Renvoyer une nouvelle liste avec seulement la première lettre
noms = ["Elie", "Tim", "Matt"]
print([nom[0] for nom in noms])

# 4. Renvoyer une nouvelle liste de toutes les valeurs paires
liste2 = [1, 2, 3, 4, 5, 6]
print([x for x in liste2 if x % 2 == 0])

# 5. Intersection de deux listes
liste3 = [1, 2, 3, 4]
liste4 = [3, 4, 5, 6]
print([x for x in liste3 if x in liste4])

# 6. Chaque mot inversé et en minuscules
mots = ["Elie", "Tim", "Matt"]
print([mot[::-1].lower() for mot in mots])

# 7. Lettres présentes dans les deux mots "first" et "third"
mot1 = "first"
mot2 = "third"
print(list(set([lettre for lettre in mot1 if lettre in mot2])))

# 8. Nombres divisibles par 12 entre 1 et 100
print([x for x in range(1, 101) if x % 12 == 0])

# 9. Voyelles supprimées de "amazing"
chaine = "amazing"
voyelles = "aeiouy"
print([lettre for lettre in chaine if lettre not in voyelles])

# 10. Liste avec la valeur [[0, 1, 2], [0, 1, 2], [0, 1, 2]]
print([[0, 1, 2] for _ in range(3)])

# 11. Matrice 10x10 avec les valeurs 0 à 9
print([[x for x in range(10)] for _ in range(10)])
```
