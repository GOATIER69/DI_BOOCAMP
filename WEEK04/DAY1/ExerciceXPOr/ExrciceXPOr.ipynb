1. Définition des termes dans le contexte de la détection de spams

Vrais positifs (TP) : Ce sont les courriels qui sont réellement des spams et que le classificateur a correctement identifiés comme spams. Dans notre exemple, 85 spams ont été bien détectés. Cela signifie que le modèle a réussi à envoyer ces courriels indésirables dans le dossier spam.

Vrais négatifs (TN) : Ce sont les courriels qui ne sont pas des spams (c’est-à-dire des courriels légitimes) et que le classificateur a correctement identifiés comme non-spams. Ici, 90 courriels légitimes ont été correctement classés et sont restés dans la boîte de réception principale.

Faux positifs (FP) : Ce sont les courriels qui sont réellement des non-spams (courriels légitimes) mais que le classificateur a incorrectement étiquetés comme spams. Dans notre exemple, 10 courriels importants ont été envoyés à tort dans le dossier spam. C’est ce qu’on appelle une « erreur de type I » : un courriel légitime est considéré comme indésirable.

Faux négatifs (FN) : Ce sont les courriels qui sont réellement des spams mais que le classificateur a incorrectement étiquetés comme non-spams. Ici, 15 spams ont atterri dans la boîte de réception principale. C’est une « erreur de type II » : un courriel indésirable n’est pas détecté et vient polluer la boîte de réception de l’utilisateur.

3. Impact d’un nombre plus élevé de faux positifs par rapport aux faux négatifs, et vice versa

Scénario A : Beaucoup de faux positifs (FP élevé) avec peu de faux négatifs (FN faible)

Cela signifie que le classificateur est très agressif : il étiquette de nombreux courriels comme spams, y compris beaucoup de courriels légitimes. Conséquences :

La précision chute fortement car de nombreux courriels prédits « spam » sont en réalité légitimes.

Le rappel reste élevé (voire parfait) car presque tous les vrais spams sont détectés.

L’exactitude peut baisser modérément si les FP sont très nombreux.

Impact utilisateur : l’utilisateur manque des courriels importants qui finissent dans le dossier spam. C’est très frustrant car il doit constamment vérifier son dossier spam. Dans un contexte professionnel, cela peut faire rater des messages clients critiques.

Scénario B : Beaucoup de faux négatifs (FN élevé) avec peu de faux positifs (FP faible)

Cela signifie que le classificateur est très prudent : il n’ose presque jamais étiqueter un courriel comme spam, sauf s’il en est absolument certain. Conséquences :

Le rappel chute fortement car de nombreux spams réels ne sont pas détectés.

La précision reste élevée (quand le modèle dit « spam », c’est presque toujours vrai).

L’exactitude peut rester élevée si les spams sont rares, mais l’utilisateur subit l’inverse : sa boîte de réception est envahie de spams.

Impact utilisateur : l’utilisateur passe son temps à supprimer manuellement des spams. Cela nuit à la productivité et à la sécurité (risque de cliquer sur un lien malveillant dans un spam non détecté).

Comparaison des deux situations et compromis

Dans la détection de spams, le coût des deux types d’erreurs n’est pas symétrique :

Un faux positif (courriel légitime en spam) coûte du temps à l’utilisateur pour retrouver le message.

Un faux négatif (spam en boîte de réception) coûte du temps de suppression et expose à des risques de sécurité (hameçonnage, malwares).

Selon l’application, on privilégiera l’un ou l’autre. Pour un utilisateur domestique, on tolère quelques faux positifs pour éviter les spams. Pour un service hospitalier, un faux négatif (spam contenant une pièce jointe malveillante) est beaucoup plus grave. Le réglage du seuil de décision du classificateur permet d’ajuster cet équilibre : en abaissant le seuil, on augmente le rappel (plus de spams détectés) mais on diminue la précision (plus de faux positifs). En relevant le seuil, on fait l’inverse. Le choix optimal dépend du coût relatif de chaque type d’erreur dans le contexte métier spécifique.

EXERCICE 2

1. Pourquoi, dans le diagnostic médical, un taux de rappel élevé est plus important qu’un taux de précision élevé

Dans un contexte de diagnostic médical, le rappel (recall) mesure la proportion de patients réellement malades que le modèle a correctement identifiés. Un rappel élevé signifie que très peu de patients malades passent inaperçus (faux négatifs faibles). La précision (precision), quant à elle, mesure la proportion de patients prédits comme malades qui le sont réellement ; une précision faible génère des faux positifs (patients sains diagnostiqués à tort comme malades).

Le rappel est plus important que la précision dans ce domaine pour plusieurs raisons.

Première raison : le coût d’un faux négatif est dramatiquement plus élevé que celui d’un faux positif. Un faux négatif (patient malade diagnostiqué sain) signifie qu’un patient ne reçoit pas le traitement nécessaire. Les conséquences peuvent être graves : progression de la maladie, complications, voire décès évitable. En cancérologie, par exemple, manquer un diagnostic précoce de tumeur maligne réduit considérablement les chances de survie. À l’inverse, un faux positif (patient sain diagnostiqué malade) entraîne du stress, des examens complémentaires, éventuellement un traitement inutile, mais le patient n’est pas en danger immédiat de mort. Le préjudice est réel mais généralement inférieur.

Deuxième raison : le principe de précaution médicale. En médecine, on applique souvent le principe « mieux vaut un excès de prudence ». Il est préférable de convoquer un patient pour des examens approfondis (suite à un faux positif) que de le renvoyer chez lui avec une maladie non détectée. Les protocoles de dépistage (mammographie, test VIH) privilégient d’ailleurs la sensibilité (rappel) pour ne pas manquer de cas, quitte à avoir des confirmations secondaires pour éliminer les faux positifs.

Troisième raison : la possibilité de gérer les faux positifs en aval. Un faux positif peut être corrigé par des examens complémentaires (biopsie, seconde analyse, imagerie plus précise). Un faux négatif, lui, n’est pas rattrapable sans une détection tardive, souvent trop tard. Le système de soin peut absorber le coût des faux positifs par des étapes de confirmation ; il ne peut pas « rattraper » un patient non diagnostiqué.

Ainsi, dans un test de dépistage du VIH, on préfère un test très sensible (rappel proche de 100 %) même si sa spécificité est imparfaite, car les faux positifs seront éliminés par un test Western blot de confirmation. Un faux négatif enverrait un patient séropositif vers des rapports non protégés et une absence de traitement.

2. Scénario où la précision devient plus importante que le rappel

Il existe des contextes où la précision prime sur le rappel. Prenons l’exemple d’un système de filtrage automatique des candidatures à un emploi (CV screening). Une entreprise reçoit des milliers de CV et utilise un classificateur pour identifier les candidats les plus prometteurs (classe positive = « candidat à convoquer »). Dans ce scénario :

Un faux positif (candidat non qualifié convoqué par erreur) coûte du temps aux recruteurs : entretien inutile, perte de productivité, frustration des équipes. Mais ce coût reste limité et gérable.

Un faux négatif (candidat excellent écarté à tort par le modèle) est très problématique car l’entreprise rate une recrue potentiellement exceptionnelle. On pourrait donc penser que le rappel est ici aussi prioritaire.

Pourtant, dans certaines variantes, la précision devient plus importante. Imaginons un système de diagnostic automatique en salle d’urgence qui alerte les médecins pour des cas critiques. Si la précision est faible, les médecins reçoivent de nombreuses fausses alertes (faux positifs). Ils finissent par ignorer le système (syndrome de l’alerte qui crie au loup), ce qui est dangereux. Ou bien prenons un système de détection de fraudes bancaires où chaque faux positif déclenche une vérification manuelle coûteuse et une friction client (carte bloquée à tort). Si la précision est trop faible, le service fraude est submergé et les clients quittent la banque. Dans ce cas, on préfère un seuil plus élevé (moins de faux positifs), quitte à accepter quelques fraudes non détectées (faux négatifs). La précision devient alors prioritaire pour des raisons opérationnelles et de satisfaction client.

Un autre exemple concret est la classification de courriels comme « important » pour un assistant personnel. Un faux négatif (courriel important classé normal) est embêtant. Mais un faux positif (courriel normal classé important) pollue la boîte de réception prioritaire et rend l’assistant inutile. Si la précision est trop faible, l’utilisateur désactive la fonction. La précision est donc plus importante que le rappel.

3. Conséquences potentielles d’une focalisation exclusive sur l’amélioration de la précision dans les ensembles de données déséquilibrés

Un ensemble de données déséquilibré est une situation où une classe (souvent la classe d’intérêt, comme la maladie ou la fraude) est très minoritaire par rapport à l’autre. Par exemple, 1 % des patients sont malades, 99 % sont sains. Si l’on se focalise exclusivement sur l’amélioration de la précision, plusieurs conséquences négatives apparaissent.

Première conséquence : chute dramatique du rappel. Pour augmenter la précision, on relève le seuil de décision du classificateur : on n’étiquette un cas comme positif qu’avec une très forte probabilité. Cela réduit les faux positifs (donc précision élevée), mais augmente mécaniquement les faux négatifs (donc rappel faible). On peut aboutir à un modèle qui ne détecte que les cas les plus évidents de maladie (les patients très symptomatiques) et rate tous les cas subtils ou à un stade précoce. Le rappel peut tomber à 10 % ou 20 %, rendant le modèle inutile en pratique car la majorité des malades sont ignorés.

Deuxième conséquence : illusion de performance. Sur un dataset déséquilibré (99 % sains, 1 % malades), un modèle qui prédit systématiquement « sain » obtient une exactitude (accuracy) de 99 %, mais une précision sur la classe malade est indéfinie (division par zéro) et un rappel de 0 %. Si l’on se focalise sur la précision, on risque d’accepter ce genre de modèle « paresseux » qui ne prédit jamais la classe minoritaire. Le modèle aura une précision artificiellement parfaite (aucun faux positif car aucune prédiction positive) mais sera totalement inutile.

Troisième conséquence : dégradation de l’équilibre métier. En cherchant à maximiser la précision, on rend le modèle extrêmement conservateur. Dans un contexte médical, cela signifie que de nombreux patients malades ne seront pas diagnostiqués (faux négatifs). Le coût humain est immense. Dans la détection de fraudes, cela laisse passer des fraudes coûteuses. Dans la modération de contenu toxique, cela laisse circuler des messages dangereux. On sacrifie la détection (rappel) au profit de la pureté des alertes (précision), ce qui contredit souvent l’objectif métier initial.

Quatrième conséquence : difficulté à comparer les modèles. Si l’on maximise la précision sans jamais regarder le rappel, on peut sélectionner un modèle qui a une précision de 99 % mais un rappel de 5 %, alors qu’un autre modèle avec une précision de 80 % et un rappel de 90 % serait bien plus utile. La focalisation exclusive sur la précision masque ce compromis fondamental.

Conclusion pratique : Dans les problèmes à classes déséquilibrées, il ne faut jamais regarder la précision isolément. On utilise des métriques adaptées comme l’AUC-PR (aire sous la courbe précision-rappel) ou le score F1, et on fixe un seuil de décision qui reflète le coût réel des faux positifs et faux négatifs dans le contexte métier. Un bon modèle équilibre précision et rappel, ou privilégie l’un ou l’autre selon les conséquences opérationnelles, mais jamais l’un au détriment total de l’autre.

EXERCICE 3

1. Différence entre validation croisée K-fold et validation croisée K-fold stratifiée. Laquelle choisir pour cette tâche et pourquoi ?

Validation croisée K-fold standard : L’ensemble de données est partitionné aléatoirement en K plis (sous-ensembles) de taille approximativement égale. Le modèle est entraîné K fois : à chaque itération, un pli différent sert d’ensemble de validation, et les K-1 plis restants servent d’ensemble d’entraînement. La performance finale est la moyenne des performances sur les K plis de validation. Cette méthode ne tient pas compte de la distribution de la variable cible.

Validation croisée K-fold stratifiée : C’est une variante qui conserve, dans chaque pli, la même proportion de chaque classe de la variable cible que dans l’ensemble complet. Autrement dit, la stratification préserve la distribution de la variable à prédire. Pour un problème de classification, on s’assure que chaque pli contient approximativement le même pourcentage d’exemples de chaque catégorie.

Laquelle choisir pour la prédiction des prix des logements ?

Pour une tâche de régression comme la prédiction des prix des logements, la variable cible est continue (le prix), et non pas catégorielle. La validation croisée stratifiée standard n’est pas directement applicable car elle est conçue pour la classification. Cependant, il existe des extensions comme la validation croisée par quantiles ou la stratification par tranches (bins) : on discrétise la variable cible en plusieurs catégories (par exemple, prix bas, moyen, élevé) et on applique une stratification sur ces tranches.

Dans la pratique pour la régression :

Si la variable cible est uniformément distribuée et qu’il n’y a pas de risque de biais, le K-fold standard est suffisant.

Si la variable cible présente des déséquilibres (peu de logements très chers, beaucoup de logements moyens), ou si certaines tranches de prix sont sous-représentées, il est préférable d’utiliser un K-fold stratifié par tranches (via StratifiedKFold après avoir créé des bins, ou StratifiedShuffleSplit). Cela évite qu’un pli contienne par hasard uniquement des logements chers, ce qui fausserait l’évaluation.

Recommandation pour ce projet : Je choisirais un K-fold stratifié par tranches de prix (par exemple, 5 tranches de prix équilibrées). La raison est que les prix des logements suivent souvent une distribution asymétrique (loi de Pareto : peu de logements très chers). Sans stratification, certains plis pourraient manquer totalement de logements à prix élevé, entraînant une mauvaise estimation de l’erreur sur ces tranches rares mais importantes. La stratification garantit que chaque pli reflète fidèlement la distribution globale des prix.

2. Description des courbes d’apprentissage et leur utilité pour comprendre les performances du modèle

Définition : Une courbe d’apprentissage est un graphique qui montre l’évolution des performances d’un modèle (généralement l’erreur ou la précision) en fonction de la taille de l’ensemble d’entraînement. On trace habituellement deux courbes sur le même graphique :

La performance sur l’ensemble d’entraînement (train score).

La performance sur l’ensemble de validation (validation score ou test score).

Utilité des courbes d’apprentissage :

Elles permettent de diagnostiquer le comportement du modèle et de répondre à des questions cruciales :

Le modèle bénéficierait-il de plus de données ? Si les deux courbes (entraînement et validation) se rapprochent d’un plateau et que l’écart entre elles est faible, ajouter plus de données n’améliorera pas significativement le modèle.

Le modèle souffre-t-il de sous-apprentissage (bias élevé) ? Si la performance sur l’entraînement est faible et que la performance sur la validation est également faible (avec un petit écart), le modèle est trop simple pour capturer la complexité des données.

Le modèle souffre-t-il de surapprentissage (variance élevée) ? Si la performance sur l’entraînement est très bonne mais que la performance sur la validation est nettement inférieure (grand écart), le modèle a mémorisé le bruit des données d’entraînement et ne généralise pas.

Aider à choisir la complexité du modèle : En traçant des courbes pour différentes complexités (par exemple, degré d’un polynôme, profondeur d’un arbre), on peut sélectionner le niveau de complexité qui minimise l’erreur de validation.

Détecter l’effet du réglage des hyperparamètres : On peut observer comment la régularisation (par exemple, le paramètre C d’une régression logistique) modifie l’écart entre les courbes.

Exemple concret pour la prédiction des prix : On trace l’erreur quadratique moyenne (MSE) sur l’entraînement et sur la validation en fonction du nombre d’exemples (de 100 à 10 000). Si l’erreur de validation continue de baisser avec plus de données, on sait qu’il vaut la peine de collecter davantage d’observations (plus de transactions immobilières). Si l’erreur de validation stagne, c’est qu’il faut améliorer le modèle (ajouter des caractéristiques, essayer un algorithme différent).

3. Implications du sous-apprentissage et du surapprentissage observés à partir des courbes d’apprentissage, et remèdes possibles

Sous-apprentissage (underfitting) – biais élevé

Apparence sur la courbe d’apprentissage :

L’erreur d’entraînement est élevée.

L’erreur de validation est également élevée, proche de l’erreur d’entraînement (faible écart).

Les deux courbes plafonnent rapidement et ne s’améliorent pas avec plus de données.

Implications : Le modèle est trop simple pour capturer la relation entre les caractéristiques et le prix des logements. Par exemple, une régression linéaire avec seulement la surface comme variable ne peut pas modéliser l’effet complexe de la localisation, du quartier, des écoles à proximité, etc. Le modèle fait des erreurs systématiques (biais élevé), et augmenter la quantité de données n’aide pas.

Remèdes possibles :

Augmenter la complexité du modèle (passer de la régression linéaire à un modèle non linéaire comme un arbre de décision, Random Forest, ou XGBoost).

Ajouter des caractéristiques pertinentes (par exemple, l’année de construction, la distance au centre-ville, le score de l’école du quartier).

Créer des caractéristiques polynomiales ou d’interaction (surface × nombre de chambres).

Réduire la régularisation (si on utilise Lasso/Ridge).

Changer d’algorithme pour un modèle plus expressif.

Sur-apprentissage (overfitting) – variance élevée

Apparence sur la courbe d’apprentissage :

L’erreur d’entraînement est très faible (parfois proche de zéro).

L’erreur de validation est beaucoup plus élevée et peut même augmenter alors que l’erreur d’entraînement continue de baisser.

Un grand écart (gap) persiste entre les deux courbes, et plus on ajoute de données, plus l’écart peut se réduire mais reste significatif.

Implications : Le modèle est trop complexe par rapport à la quantité de données disponibles. Il a mémorisé le bruit et les spécificités de l’ensemble d’entraînement, y compris les erreurs de mesure ou les anomalies, plutôt que d’apprendre le schéma général. Par exemple, un arbre de décision très profond (max_depth=30) sur 500 logements va créer des règles absurdes (« si la couleur de la porte est bleue et qu’il y a un arbre dans le jardin, alors prix = 500k »). En production, le modèle se comporte mal.

Remèdes possibles :

Augmenter la taille de l’ensemble d’entraînement (collecter plus de transactions immobilières) – c’est souvent la solution la plus efficace.

Réduire la complexité du modèle (diminuer la profondeur des arbres, réduire le nombre de caractéristiques, utiliser un modèle plus simple).

Appliquer des techniques de régularisation (L1/L2 pour la régression, min_samples_split, max_features pour les arbres, dropout pour les réseaux de neurones).

Utiliser la validation croisée pour le réglage des hyperparamètres et sélectionner la complexité optimale.

Ajouter du bruit aux données d’entraînement (technique de regularisation implicite).

Faire du feature selection pour éliminer les caractéristiques non informatives ou redondantes.

Utiliser l’arrêt précoce (early stopping) dans les modèles itératifs (XGBoost, réseaux de neurones).

Cas idéal : Les courbes d’entraînement et de validation sont proches l’une de l’autre, toutes deux à un niveau de faible erreur, et elles se stabilisent avec l’ajout de données. Cela indique un bon équilibre entre biais et variance.

Application à la prédiction des prix des logements : Un surapprentissage typique serait un modèle XGBoost très profond (max_depth=15) avec un faible taux d’apprentissage mais trop d’estimateurs. Les courbes montreraient une MSE d’entraînement proche de 0, mais une MSE de validation élevée. Les remèdes seraient de réduire la profondeur (max_depth=5-7), augmenter la régularisation (gamma, lambda), utiliser l’arrêt précoce, ou augmenter le nombre d’exemples (plus de transactions). Un sous-apprentissage typique serait une régression linéaire simple sans prise en compte des interactions : les deux courbes seraient élevées et plates. Le remède serait d’ajouter des caractéristiques non linéaires ou de passer à Random Forest.

EXERCICE 4

1. Pourquoi l’utilisation de la précision (accuracy) comme critère d’évaluation pourrait être trompeuse dans ce scénario

L’exactitude (accuracy) mesure la proportion de prédictions correctes (TP + TN) sur l’ensemble total des prédictions. Dans un contexte de déséquilibre extrême (2 % de positifs, 98 % de négatifs), un modèle peut obtenir une exactitude très élevée sans jamais détecter le moindre malade.

Exemple chiffré : Supposons un ensemble de test de 1 000 patients. Il y a 20 malades (2 %) et 980 sains (98 %). Un modèle « naïf » qui prédit systématiquement « sain » pour tous les patients obtient :

TP = 0 (aucun malade détecté)

TN = 980 (tous les sains correctement classés)

FP = 0

FN = 20 (tous les malades sont des faux négatifs)

L’exactitude est : (TP + TN) / total = (0 + 980) / 1000 = 0,98 soit 98 % d’exactitude. Pourtant, ce modèle est totalement inutile car il ne détecte aucun cas de maladie. Il laisse passer tous les patients malades, ce qui est catastrophique dans un contexte médical.

Pourquoi c’est trompeur : L’exactitude ne distingue pas entre les classes. Elle est dominée par la classe majoritaire (les sains). Un modèle peut être excellent sur la classe majoritaire et exécrable sur la classe minoritaire (celle qui nous intéresse vraiment) tout en affichant une exactitude impressionnante. Un analyste non averti pourrait croire que son modèle est performant alors qu’il est inutilisable en pratique. C’est pourquoi, dans les problèmes déséquilibrés, l’exactitude est souvent abandonnée au profit d’autres métriques.

2. Importance de la précision et du rappel dans le contexte de cet ensemble de données déséquilibré

Le rappel (recall) est crucial : Le rappel mesure la proportion de vrais malades correctement identifiés : TP / (TP + FN). Dans une maladie rare, ne pas détecter un patient malade (faux négatif) a des conséquences potentiellement graves : absence de traitement, progression de la maladie, risque de transmission (pour les maladies infectieuses), voire décès. Pour un test de dépistage, on veut un rappel aussi proche de 100 % que possible, quitte à générer des faux positifs. Un mauvais rappel (par exemple 50 %) signifierait que la moitié des patients atteints repartent avec un diagnostic erroné. C’est inacceptable. Le rappel répond à la question : « Parmi tous les patients vraiment malades, combien notre modèle a-t-il sauvés ? »

La précision (precision) est également importante, mais dans un second temps : La précision mesure la proportion de prédictions « malade » qui sont correctes : TP / (TP + FP). Une précision faible signifie que le modèle émet beaucoup de fausses alertes (faux positifs). Dans un contexte médical, les faux positifs ont des coûts : anxiété pour le patient, examens complémentaires invasifs (biopsie, coloscopie), surtraitement, engorgement du système de soin, et coûts financiers. Cependant, ces conséquences sont généralement moins graves qu’un faux négatif. Une précision très faible (par exemple 1 %) rendrait le modèle inutilisable car chaque prédiction positive serait presque toujours fausse, et on soumettrait des milliers de patients sains à des investigations inutiles.

L’équilibre : Dans les maladies rares, on privilégie généralement le rappel, mais avec une contrainte minimale sur la précision. On tolère quelques faux positifs pour capturer presque tous les vrais positifs. On peut par exemple fixer un objectif de rappel ≥ 95 %, puis chercher à maximiser la précision sous cette contrainte. Le score F1 (moyenne harmonique) est utile pour comparer des modèles, mais attention : avec un déséquilibre extrême, le F1 peut encore être trompeur car il est asymétrique. L’aire sous la courbe précision-rappel (AUC-PR) est souvent préférée à l’AUC-ROC pour les problèmes très déséquilibrés.

3. Stratégies pour évaluer et améliorer les performances du modèle face au déséquilibre

Stratégies d’évaluation :

Abandonner l’accuracy : Ne pas l’utiliser comme métrique principale. Utiliser plutôt le rappel, la précision, le F1, et surtout l’AUC-PR (Precision-Recall AUC). La courbe précision-rappel est beaucoup plus informative que la courbe ROC quand la classe positive est rare.

Matrice de confusion : Toujours examiner la matrice de confusion pour visualiser les TP, TN, FP, FN. Elle révèle immédiatement le comportement sur la classe minoritaire.

Seuils de décision ajustables : Ne pas se limiter au seuil par défaut de 0,5. Tracer la courbe précision-rappel pour différents seuils et choisir le seuil qui maximise un critère métier (par exemple, rappel ≥ 95 % avec la précision la plus élevée possible).

Validation croisée stratifiée : Utiliser une validation croisée stratifiée pour garantir que chaque pli contient environ 2 % de malades (sinon, un pli pourrait n’avoir aucun malade et fausser l’évaluation).

Stratégies pour améliorer le modèle (techniques de rééchantillonnage et algorithmiques) :

Sur-échantillonnage de la classe minoritaire (oversampling) : Dupliquer aléatoirement les exemples de la classe malade pour équilibrer les classes. Une version plus avancée est SMOTE (Synthetic Minority Over-sampling Technique) qui crée des exemples synthétiques en interpolant entre des malades proches. Attention : SMOTE peut créer des exemples irréalistes dans certains espaces de caractéristiques.

Sous-échantillonnage de la classe majoritaire (undersampling) : Réduire aléatoirement le nombre de patients sains pour équilibrer. Risque : perdre des informations utiles. Des méthodes comme NearMiss ou Tomek links permettent un sous-échantillonnage plus intelligent.

Pondération des classes (class weights) : Assigner un poids plus élevé aux erreurs sur la classe minoritaire. Dans la régression logistique ou les forêts aléatoires, on peut utiliser le paramètre class_weight='balanced' qui calcule automatiquement des poids inversement proportionnels aux fréquences des classes. Par exemple, un faux négatif (malade non détecté) coûtera 49 fois plus qu’un faux positif (car 98/2 = 49).

Utiliser des algorithmes robustes au déséquilibre : XGBoost permet de régler scale_pos_weight (ratio du nombre de négatifs sur les positifs). Les forêts aléatoires peuvent utiliser un échantillonnage équilibré (bootstrap avec remplacement équilibré).

Seuil de décision personnalisé : Après entraînement, ajuster le seuil de classification. Au lieu de 0,5, on peut le baisser (par exemple à 0,3) pour augmenter le rappel (plus de prédictions positives, donc plus de malades détectés, mais aussi plus de faux positifs). On détermine le seuil optimal en maximisant une métrique métier (ex : score F1 ou un coût personnalisé).

Collecte de plus de données de la classe minoritaire : Si possible, aller chercher activement plus d’exemples de malades (par exemple via des campagnes de dépistage ciblées). C’est souvent la solution la plus efficace, mais aussi la plus coûteuse.

Utiliser des métriques alternatives pour le réglage : Lors de la recherche d’hyperparamètres (GridSearchCV), optimiser non pas l’accuracy mais l’AUC-PR ou le F1. On peut paramétrer GridSearchCV avec scoring='f1' ou scoring='average_precision'.

Méthodes avancées : Les approches basées sur les anomalies (one-class classification) ou l’apprentissage sensible aux coûts (cost-sensitive learning) sont également pertinentes pour les déséquilibres extrêmes.

Exemple concret pour la maladie rare : Supposons qu’après entraînement, le modèle ait un rappel de 60 % (il détecte 12 malades sur 20) et une précision de 4 % (sur 100 prédictions « malade », seulement 4 sont correctes). On applique SMOTE pour générer des malades synthétiques. On abaisse le seuil de 0,5 à 0,2. Le nouveau seuil fait passer le rappel à 90 % (18/20 malades détectés) mais la précision tombe à 2 % (il y a beaucoup de faux positifs). On juge si ce compromis est acceptable : détecter 18 malades sur 20 avec 900 faux positifs (pour 1 000 patients) générera 900 examens complémentaires inutiles. Le médecin peut décider que c’est trop, et on remontera le seuil à 0,35 pour obtenir un rappel de 80 % avec une précision de 10 %. La décision finale dépend du coût relatif d’un examen inutile versus celui d’une maladie non détectée.

EXERCICE 5

1. Effet du passage du seuil de 0,5 à 0,7 sur la précision et le rappel

Seuil par défaut (0,5) : Le modèle classe un client comme étant à risque de défaut dès que sa probabilité prédite atteint 50 %.

Seuil relevé (0,7) : Le modèle devient plus exigeant : il ne prédit un défaut que pour les clients dont la probabilité est très élevée (≥ 70 %).

Effet sur le rappel (recall) : Le rappel mesure la proportion de vrais défauts correctement identifiés. En relevant le seuil à 0,7, on prédit beaucoup moins de positifs. Parmi les clients qui feront réellement défaut, certains avaient une probabilité modérée (par exemple entre 0,5 et 0,7). Ils ne seront plus détectés et deviendront des faux négatifs. Le nombre de vrais positifs (TP) diminue tandis que les faux négatifs (FN) augmentent. Par conséquent, le rappel diminue (parfois très significativement). Le modèle devient plus sélectif et rate une partie des mauvais payeurs.

Effet sur la précision (precision) : La précision mesure la proportion de clients prédits comme « défaut » qui le sont réellement. En relevant le seuil, on élimine de nombreuses prédictions positives à risque modéré. Parmi les clients dont la probabilité dépasse 0,7, une proportion plus élevée sont effectivement de mauvais payeurs. Les faux positifs (clients solvables prédits à tort comme défaut) diminuent fortement. La précision augmente généralement. Le modèle est plus fiable : lorsqu’il dit qu’un client va faire défaut, il a plus souvent raison.

Résumé : Seuil 0,7 → rappel plus faible (on détecte moins de défauts réels) mais précision plus élevée (les alertes de défaut sont plus fiables). On sacrifie la détection au profit de la pureté des alertes.

Exemple chiffré : Supposons qu’avec le seuil 0,5, on ait un rappel de 85 % et une précision de 60 %. Avec un seuil 0,7, le rappel pourrait chuter à 50 % (beaucoup de défauts réels non détectés car leur probabilité était entre 0,5 et 0,7), tandis que la précision pourrait monter à 80 % (les clients prédits comme défauts sont très risqués).

2. Conséquences potentielles d’un seuil trop élevé ou trop bas dans le contexte bancaire

Seuil trop bas (par exemple 0,2 ou 0,1) : Le modèle est très sensible et prédit un défaut pour de nombreux clients, même avec une faible probabilité.

Conséquences :

Rappel très élevé : presque tous les vrais défauts sont détectés. La banque identifie quasiment tous les mauvais payeurs potentiels.

Précision très faible : beaucoup de faux positifs. De nombreux clients solvables (qui rembourseront leur prêt) sont classés à tort comme risqués.

Impact métier : La banque va refuser des prêts à des clients qui auraient pourtant bien remboursé. Perte de chiffre d’affaires (intérêts non perçus), mécontentement client, dégradation de l’image de marque. De plus, les clients solvables refusés iront chez un concurrent. La banque peut également appliquer des taux d’intérêt très élevés à ces clients « faux positifs », ce qui est injuste et peut poser des problèmes réglementaires (discrimination indirecte). Coût opérationnel : traiter manuellement de nombreux dossiers refusés à tort.

Seuil trop élevé (par exemple 0,8 ou 0,9) : Le modèle est extrêmement prudent et ne prédit un défaut que pour les cas les plus évidents.

Conséquences :

Précision très élevée : quand la banque refuse un prêt pour risque de défaut, elle a presque toujours raison. Les rares clients qu’elle qualifie de risqués le sont effectivement.

Rappel très faible : de nombreux vrais défauts ne sont pas détectés. La banque accorde des prêts à des clients qui feront défaut, croyant qu’ils sont solvables.

Impact métier : Pertes financières directes sur les prêts non remboursés. Un défaut de paiement peut entraîner une perte partielle ou totale du capital prêté. Pour une banque, un seul prêt en défaut sur un gros montant peut anéantir les bénéfices réalisés sur des dizaines de prêts sains. L’exposition au risque de crédit augmente dangereusement. De plus, la banque doit provisionner des fonds pour couvrir les pertes attendues, ce qui réduit ses capacités de prêt.

Compromis optimal : La banque doit équilibrer le coût d’un faux positif (perte d’un bon client, manque à gagner sur les intérêts) et le coût d’un faux négatif (perte du capital prêté en cas de défaut). En général, le coût d’un défaut non détecté (faux négatif) est beaucoup plus élevé que le coût d’un refus de prêt à un bon client (faux positif). On tolérera donc un seuil plutôt bas pour favoriser un rappel élevé, mais pas trop bas pour ne pas générer trop de faux positifs. Le seuil de 0,5 est un point de départ ; les banques ajustent souvent vers 0,3-0,4 pour être plus prudentes.

3. Utilisation des courbes ROC et de l’AUC pour trouver le seuil optimal

Courbe ROC (Receiver Operating Characteristic) : La courbe ROC représente graphiquement les performances d’un classifieur pour tous les seuils possibles. Elle trace le taux de vrais positifs (rappel) sur l’axe des ordonnées en fonction du taux de faux positifs (FPR = FP / (FP + TN)) sur l’axe des abscisses. Chaque point de la courbe correspond à un seuil différent. Le point (0,1) représente un classifieur parfait (aucun faux positif, tous les vrais positifs détectés). La diagonale (ligne de 45 degrés) représente un classifieur aléatoire.

AUC (Area Under the Curve) : L’AUC est la surface sous la courbe ROC, comprise entre 0 et 1. Un AUC de 0,5 indique un modèle aléatoire ; un AUC de 1 indique un modèle parfait. L’AUC mesure la capacité du modèle à distinguer les deux classes indépendamment de tout seuil. Elle est utile pour comparer des modèles entre eux.

Comment la courbe ROC aide à choisir le seuil optimal :

Identifier le compromis souhaité : Sur la courbe ROC, chaque seuil correspond à un couple (FPR, TPR). On peut parcourir la courbe et choisir le seuil qui maximise une métrique métier spécifique.

Maximiser le score F1 : Pour chaque seuil, on peut calculer le score F1 (moyenne harmonique de la précision et du rappel) et choisir le seuil qui donne le F1 le plus élevé. Bien que le F1 ne soit pas directement tracé sur la courbe ROC, on peut le calculer à partir des valeurs de précision et de rappel dérivées de la matrice de confusion pour chaque seuil.

Minimiser un coût métier personnalisé : La banque peut définir une fonction de coût : Coût = C_FP * FP + C_FN * FN, où C_FP est le coût d’un faux positif (refus injustifié) et C_FN le coût d’un faux négatif (défaut non détecté). En parcourant tous les seuils, on trouve celui qui minimise le coût total. Par exemple, si C_FN = 10 * C_FP (un défaut non détecté coûte 10 fois plus cher qu’un refus injustifié), on choisira un seuil plus bas pour favoriser le rappel.

Utiliser la courbe de Youden : L’indice de Youden est J = TPR - FPR. Le seuil qui maximise J est souvent un bon compromis équilibré (maximise la différence entre le taux de vrais positifs et le taux de faux positifs). C’est un point de départ neutre avant d’intégrer des coûts métier.

Tracer la précision et le rappel en fonction du seuil : Une méthode plus directe que la courbe ROC est de tracer séparément la précision et le rappel en fonction du seuil. On cherche le seuil où les deux courbes se croisent (si l’on souhaite un équilibre) ou le seuil qui maximise une métrique composée.

Mise en garde sur l’AUC dans les contextes déséquilibrés : L’AUC-ROC peut être trop optimiste lorsque le déséquilibre des classes est extrême (par exemple 2 % de défauts). Dans ce cas, l’AUC-PR (aire sous la courbe précision-rappel) est beaucoup plus informative, car elle se concentre sur la classe minoritaire. Pour la prédiction des défauts de prêt, qui est souvent déséquilibrée (peu de défauts), je recommande d’utiliser l’AUC-PR plutôt que l’AUC-ROC pour évaluer les modèles.

Procédure pratique pour la banque :

Entraîner le modèle et obtenir les probabilités de défaut pour chaque client de l’ensemble de validation.

Trier les clients par probabilité décroissante.

Pour chaque seuil possible (ou pour une grille de seuils), calculer la matrice de confusion, puis la précision, le rappel, et le coût métier total.

Choisir le seuil qui minimise le coût total (ou qui satisfait une contrainte réglementaire, par exemple un rappel minimal de 80 %).

Valider ce seuil sur un ensemble de test distinct.

Réévaluer périodiquement car les coûts et la distribution des défauts peuvent évoluer (concept drift).
