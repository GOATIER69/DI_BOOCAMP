2) Discuss: How did the shift to modern NLP impact scalability and efficiency?

The transition from traditional to modern NLP fundamentally transformed both scalability and efficiency by shifting the bottleneck from human expertise to computational power. First, traditional models required manual feature engineering for every new task, which did not scale with data volume; modern models, however, learn representations directly from raw text, allowing them to scale effortlessly with the size of available corpora. Second, the Transformer architecture leverages massive hardware parallelism (GPUs) to process sequences in parallel, dramatically accelerating training on billions of tokens compared to the sequential nature of older recurrent models. Third, the paradigm of pre-training and fine-tuning enables extraordinary task scalability: a single pre-trained model can be adapted to dozens of downstream tasks with minimal labeled data, eliminating the need to train from scratch each time. Fourth, transfer learning dramatically improves human efficiency, as practitioners no longer need to design task-specific features but simply fine-tune with a few hundred examples for high performance. Fifth, while modern models achieve unparalleled qualitative efficiency (beating humans on many benchmarks), they create a new scalability challenge in deployment, as their immense size demands sophisticated infrastructure for serving at scale. Finally, the ability to use few-shot and zero-shot prompting with models like GPT has further revolutionized efficiency, allowing non-experts to solve NLP tasks without any training or labeled data, something unthinkable in the traditional paradigm. Ultimately, modern NLP trades computational efficiency (higher cost) for massive gains in human efficiency and task adaptability, enabling applications that were impossible just a decade ago.


## 🌟 Exercise 2: LLM Architecture and Application Scenarios
For each, describe (a) core architectural differences, (b) a real application, (c) why it fits.

### BERT
| Caractéristique | Description |
|:---|:---|
| **Directionnalité** | **Bidirectionnel** - Lit le texte dans les deux sens (gauche→droite et droite→gauche) simultanément. Chaque token peut "voir" tous les autres tokens de la phrase. |
| **Objectif de pré-formation** | **MLM (Masked Language Model)** - 15% des tokens sont masqués aléatoirement, le modèle doit les prédire en utilisant le contexte des deux côtés. |
| **Architecture** | **Encoder-only** - Utilise uniquement la partie encodeur du Transformer. Pas de mécanisme de décodage auto-régressif. |
| **Objectif secondaire** | **NSP (Next Sentence Prediction)** - Prédit si deux phrases se suivent logiquement. |
| **Type d'attention** | **Self-attention bidirectionnelle** - Chaque token attend tous les autres tokens de la séquence. |

### 🎯 Application concrète : Analyse de sentiment dans les avis clients

**Scénario :** Une plateforme de e-commerce souhaite analyser automatiquement des millions d'avis clients pour déterminer le sentiment (positif, négatif, neutre) et identifier des aspects spécifiques (qualité du produit, service client, prix, livraison).

### ✅ Pourquoi BERT est particulièrement adapté

| Avantage | Explication |
|:---|:---|
| **Compréhension contextuelle fine** | L'architecture bidirectionnelle capture les nuances du langage naturel. Ex: *"Le produit est bon, mais le service client est mauvais"* - BERT comprend que "bon" et "mauvais" se réfèrent à des aspects différents. |
| **Gestion des ambiguïtés** | Résout les ambiguïtés sémantiques. Ex: *"Ce film est vraiment nul"* - BERT comprend "nul" comme négatif dans ce contexte. |
| **Classification fine** | La couche [CLS] (classification token) est spécifiquement conçue pour les tâches de classification. |
| **Transfert learning efficace** | Fine-tuning possible avec très peu d'exemples étiquetés (~100-500 avis). |
| **Robustesse** | Maintient des performances robustes face aux fautes d'orthographe, jargon, émojis et formulations non standard. |


### GPT

| Caractéristique | Description |
|:---|:---|
| **Directionnalité** | **Unidirectionnel (causal)** - Lit le texte de gauche à droite uniquement. Chaque token ne peut "voir" que les tokens précédents (contexte passé). |
| **Objectif de pré-formation** | **CLM (Causal Language Modeling)** - Prédit le prochain token dans une séquence en se basant uniquement sur les tokens précédents. |
| **Architecture** | **Decoder-only** - Utilise uniquement la partie décodeur du Transformer avec une attention causale (masquée). |
| **Méthode de génération** | **Auto-régressif** - Génère du texte token par token, chaque nouveau token est basé sur tous les tokens précédemment générés. |
| **Type d'attention** | **Self-attention causale** - Chaque token ne peut attendre que les tokens à sa gauche (masquage des tokens futurs). |
| **Flexibilité** | **Few-shot / Zero-shot learning** - Peut être utilisé avec des prompts sans fine-tuning grâce à la compréhension des instructions. |

### 🎯 Application concrète : Assistant virtuel de rédaction de contenu marketing

**Scénario :** Une agence de marketing digital souhaite un assistant IA capable de générer des descriptions de produits, des articles de blog, des emails de prospection et des posts sur les réseaux sociaux, tout en adaptant le ton (formel, décontracté, persuasif) et la longueur selon les besoins.

### ✅ Pourquoi GPT est particulièrement adapté

| Avantage | Explication |
|:---|:---|
| **Génération de texte fluide et cohérente** | L'architecture auto-régressive est naturellement conçue pour la génération de texte. Produit des séquences cohérentes où chaque mot s'enchaîne logiquement. |
| **Adaptabilité contextuelle** | Ajuste son style en fonction des instructions dans le prompt. Ex: *"Rédige un email professionnel"* vs *"Écris un post Instagram décontracté"*. |
| **Gestion de la longueur variable** | Peut générer du texte de n'importe quelle longueur (50 à 1000+ mots), idéal pour différents formats de contenu marketing. |
| **Few-shot learning intégré** | Peut apprendre à imiter un style à partir de quelques exemples dans le prompt (in-context learning), sans ré-entraînement. |
| **Continuité thématique** | Maintient le sujet sur de longs textes, créant des articles de blog complets et structurés. |
| **Versatilité** | Un seul modèle GPT gère tous les types de contenu (descriptions, blogs, emails, réseaux sociaux). |


### T5
| Caractéristique | Description |
|:---|:---|
| **Directionnalité** | **Encoder bidirectionnel + Decoder unidirectionnel** - Combine les deux approches : l'encodeur lit l'entrée dans les deux sens, le décodeur génère la sortie de manière auto-régressive. |
| **Objectif de pré-formation** | **Span Corruption** - Des portions de texte (spans) sont masquées, le modèle prédit les tokens manquants. Plus flexible que le MLM standard. |
| **Architecture** | **Encoder-Decoder complet** - Utilise l'architecture Transformer complète avec encodeur et décodeur, similaire au modèle original. |
| **Unification des tâches** | **Text-to-text** - Toutes les tâches NLP sont reformulées comme des problèmes de conversion texte→texte. |
| **Type d'attention** | **Attention croisée (Cross-attention)** - Le décodeur utilise une attention croisée sur les représentations de l'encodeur. |
| **Pré-formation massive** | **C4 (Colossal Clean Crawled Corpus)** - Pré-entraîné sur 750 GB de données web nettoyées. |

### 🎯 Application concrète : Système de résumé automatique de rapports financiers

**Scénario :** Une banque d'investissement reçoit quotidiennement des centaines de rapports financiers, analyses de marché et rapports annuels d'entreprises (10-K, 10-Q). Elle souhaite un système capable de résumer automatiquement ces documents en 1-2 paragraphes concis pour permettre aux analystes d'identifier rapidement les informations critiques.

### ✅ Pourquoi T5 est particulièrement adapté

| Avantage | Explication |
|:---|:---|
| **Architecture Encoder-Decoder optimisée** | L'encodeur capture la compréhension complète du document source (bidirectionnel), tandis que le décodeur génère un résumé fluide et bien structuré. |
| **Gestion de la compression d'information** | Excelle à "compresser" un long texte (5000 mots) en une sortie courte (100 mots). L'attention croisée permet de se concentrer sur les parties importantes. |
| **Flexibilité text-to-text** | Un seul modèle peut être fine-tuné pour différents styles de résumé : exécutif, détaillé, orienté investisseur, orienté risque. |
| **Capacité de reformulation** | Génère des phrases originales qui reformulent l'information (abstractive summarization), produisant des résumés plus naturels. |
| **Gestion des documents longs** | Peut traiter des documents de plusieurs pages et en extraire l'essentiel grâce à des mécanismes d'attention efficaces. |
| **Résilience aux données complexes** | Comprend le jargon technique, les chiffres, les abréviations et les formulations juridiques complexes des rapports financiers. |
| **Adaptabilité multilingue** | Peut être adapté pour résumer des rapports en plusieurs langues (français, anglais, allemand, etc.). |

## 1️⃣ Cinq principaux avantages des modèles pré-entraînés

### 1. Généralisation améliorée

**Explication :** Les modèles pré-entraînés sur des corpus massifs et diversifiés (comme tout Wikipedia, Common Crawl ou des milliards de pages web) développent une compréhension profonde et large de la structure du langage, de la grammaire, des faits généraux, des nuances culturelles et des relations sémantiques.

**En pratique :**
- Un modèle pré-entraîné sait que "Paris est la capitale de la France" même avant d'être fine-tuné.
- Il comprend les relations entre les mots (synonymes, antonymes, hyperonymes) et les structures grammaticales complexes.
- Cette connaissance générale lui permet de mieux interpréter des textes dans des domaines variés sans avoir besoin d'être ré-entraîné de zéro.

**Exemple concret :** Un BERT pré-entraîné peut comprendre une phrase comme *"Le chat poursuivait la souris"* et *"La souris était poursuivie par le chat"* comme exprimant la même relation malgré des structures syntaxiques différentes, grâce à sa compréhension générale du langage.

---

### 2. Besoin réduit de données étiquetées

**Explication :** Les modèles pré-entraînés ont déjà appris des représentations riches du langage. Par conséquent, pour les adapter à une tâche spécifique, ils n'ont besoin que d'une fraction des données étiquetées qui seraient nécessaires pour entraîner un modèle de zéro.

**En pratique :**
- Un modèle de zéro peut nécessiter 50 000 à 100 000 exemples étiquetés.
- Un modèle pré-entraîné peut atteindre des performances similaires avec seulement 500 à 2 000 exemples.
- Dans certains cas (few-shot learning), quelques dizaines d'exemples suffisent.

**Exemple concret :** Pour une analyse de sentiment dans un domaine niche (avis médicaux), un BERT fine-tuné sur 1 000 avis étiquetés peut surpasser un modèle SVM entraîné sur 50 000 avis. Cette réduction des besoins en données est critique pour les domaines où l'étiquetage manuel est coûteux ou nécessite des experts (médecine, droit, finance).

**Impact économique :** Réduction spectaculaire des coûts d'annotation, permettant aux petites équipes et startups de développer des solutions NLP performantes sans budgets d'annotation massifs.

---

### 3. Réglage fin (Fine-tuning) plus rapide

**Explication :** Puisque le modèle pré-entraîné a déjà convergé vers une représentation utile du langage, le fine-tuning pour une tâche spécifique nécessite beaucoup moins d'itérations et d'époques d'entraînement.

**En pratique :**
- Un modèle entraîné de zéro sur 50 000 exemples peut nécessiter 30-50 époques sur plusieurs jours (voire semaines) avec des GPUs.
- Un modèle pré-entraîné fine-tuné sur 1 000 exemples peut converger en 3-5 époques en quelques heures.

**Exemple concret :**
- **Sans pré-formation :** Entraîner un Transformer de zéro sur un dataset de classification peut prendre 72 heures sur 4 GPUs.
- **Avec pré-formation :** Le même modèle fine-tuné peut atteindre des résultats supérieurs en seulement 2 heures sur un seul GPU.

**Avantage opérationnel :** Permet des cycles de développement itératifs rapides. Les équipes peuvent expérimenter plusieurs architectures, hyperparamètres et stratégies de fine-tuning sans délais prohibitifs, accélérant significativement le time-to-market.

---

### 4. Apprentissage par transfert (Transfer Learning)

**Explication :** L'apprentissage par transfert permet de réutiliser les connaissances acquises sur une tâche (ou un domaine) pour améliorer les performances sur une autre tâche (ou domaine) connexe.

**En pratique :**
- La connaissance du langage général (pré-formation) est transférée à une tâche spécifique (fine-tuning).
- Cette connaissance peut aussi être transférée d'un domaine à un autre (ex: des actualités générales à des documents juridiques).
- Les couches inférieures du modèle (qui capturent la syntaxe et la grammaire) sont souvent réutilisées telles quelles, seules les couches supérieures étant adaptées.

**Exemple concret :** Un modèle pré-entraîné sur des articles Wikipedia peut être fine-tuné pour :
1. L'analyse de sentiment dans les tweets (domaine informel)
2. La classification de documents juridiques (domaine technique)
3. La reconnaissance d'entités nommées en médecine (domaine spécialisé)

**Avantage stratégique :** L'apprentissage par transfert démocratise l'accès au NLP avancé. Une seule pré-formation coûteuse peut bénéficier à des milliers de tâches différentes via des fine-tunings peu coûteux.

---

### 5. Robustesse améliorée

**Explication :** L'exposition à des quantités massives de textes variés rend les modèles pré-entraînés plus résistants aux variations linguistiques, aux bruits, aux fautes d'orthographe, aux formulations non standards et aux variations stylistiques.

**En pratique :**
- Le modèle a vu des millions de variations de phrases similaires pendant la pré-formation.
- Il apprend à se concentrer sur les motifs sémantiques plutôt que sur les variations superficielles.
- Il développe une certaine "immunité" au bruit et aux perturbations.

**Exemple concret :** Un modèle pré-entraîné peut comprendre correctement :
- *"j é tro bo"* (texte avec fautes et abréviations)
- *"This product is lit! 🔥"* (langage informel avec émojis)
- *"Le service client? Une catastrophe."* (construction elliptique)
- *"Franchement? Bof."* (langage familier)

**Bénéfice métier :** Les applications NLP deviennent viables dans des environnements réels où les données sont rarement parfaites (réseaux sociaux, retranscriptions audio, OCR, saisies utilisateur). La robustesse réduit également les besoins en prétraitement et nettoyage des données.

## 2️⃣ Problèmes éthiques potentiels de la pré-formation sur des données massives

### A. Biais (Bias) et discrimination

#### 🔴 Problème
Les modèles pré-entraînés apprennent et amplifient les biais présents dans les données d'entraînement (textes du web), qui reflètent les préjugés sociétaux historiques.

**Exemples concrets :**

| Type de biais | Exemple | Impact |
|:---|:---|:---|
| **Biais de genre** | "Infirmière" associé aux femmes, "ingénieur" aux hommes | Renforce les stéréotypes de genre dans les recrutements automatisés |
| **Biais racial** | Associations négatives avec certains noms ou dialectes | Discrimination dans les systèmes de justice prédictive |
| **Biais culturel** | Vision occidentalocentrée du monde | Mauvaise compréhension des contextes non-occidentaux |
| **Biais socio-économique** | Langage des classes aisées sur-représenté | Modèles inefficaces pour les populations défavorisées |

**Exemple impactant :** Un système de recrutement basé sur un LLM pourrait systématiquement favoriser les candidats masculins pour des postes techniques parce que les données d'entraînement associent "programmeur" aux pronoms masculins.

---

### B. Désinformation et hallucinations

#### 🔴 Problème
Les LLM peuvent générer des informations fausses ou trompeuses avec une grande confiance ("hallucinations"), et être utilisés pour créer et diffuser de la désinformation à grande échelle.

**Exemples concrets :**

| Problème | Description | Impact |
|:---|:---|:---|
| **Hallucinations** | Le modèle invente des faits, citations, événements | Propager de fausses informations comme vraies |
| **Fake news à grande échelle** | Génération automatique d'articles mensongers | Manipulation de l'opinion publique |
| **Deepfakes textuels** | Imitation du style d'écriture d'une personne | Usurpation d'identité, fraude |
| **Amplification virale** | Désinformation produite en masse rapidement | Difficile à contrôler et corriger |

**Exemple impactant :** Un LLM peut générer un article scientifique convaincant avec des références inventées, qui pourrait être utilisé pour semer le doute sur la science du climat ou la sécurité des vaccins.

---

### C. Utilisation abusive et malveillante

#### 🔴 Problème
Les capacités avancées des LLM peuvent être détournées pour des activités malveillantes ou nuisibles.

**Exemples concrets :**

| Type d'abus | Application | Risque |
|:---|:---|:---|
| **Hameçonnage (Phishing)** | Emails personnalisés et convaincants | Escroqueries financières |
| **Impersonnation** | Copie du style d'écriture d'une personne | Fraude, chantage |
| **Contenu toxique** | Génération automatisée de discours de haine | Radicalisation, harcèlement |
| **Triche académique** | Génération d'essais, résolution d'examens | Intégrité académique compromise |
| **Désinformation politique** | Campagnes de manipulation à grande échelle | Menace pour les processus démocratiques |

**Exemple impactant :** Des cybercriminels pourraient utiliser un LLM pour générer des milliers d'emails de phishing personnalisés en quelques minutes, chacun adapté au destinataire, rendant la détection beaucoup plus difficile.

---

### D. Vie privée et données personnelles

#### 🔴 Problème
Les données d'entraînement contiennent souvent des informations personnelles qui peuvent être mémorisées et réémises par le modèle.

**Exemples concrets :**

| Problème | Description | Risque |
|:---|:---|:---|
| **Mémorisation** | Le modèle "se souvient" de données personnelles | Exposition involontaire d'informations privées |
| **Ré-identification** | Combinaison de fragments pour identifier une personne | Violation du RGPD et de la vie privée |
| **Données médicales** | Inférences sur la santé à partir de textes | Discrimination, stigmatisation |
| **Données de localisation** | Inférences sur les habitudes de vie | Surveillance, tracking |

**Exemple impactant :** Un LLM pourrait, en réponse à une requête générique, générer un texte incluant un numéro de téléphone ou une adresse email présente dans ses données d'entraînement.

---

### E. Conséquences environnementales

#### 🔴 Problème
L'entraînement des LLM géants consomme d'énormes quantités d'énergie avec des émissions de CO2 significatives.

**Exemples concrets :**

| Modèle | CO₂ émis | Équivalent |
|:---|:---|:---|
| **GPT-3** | ~500 tonnes | 125 allers-retours Paris-New York en avion |
| **BERT Base** | ~1.4 tonnes | Un an de consommation électrique d'un foyer américain |
| **Training d'un LLM** | Plusieurs centaines de tonnes | À vie de 5-10 personnes en moyenne |

**Impact :** La course aux modèles toujours plus grands (milliards → trillions de paramètres) aggrave le bilan carbone et pose des questions de durabilité environnementale.

## 3️⃣ Stratégies d'atténuation des risques éthiques

### A. Pour les biais et la discrimination

| Stratégie | Description | Exemple |
|:---|:---|:---|
| **Audit des données** | Analyse systématique des données pour identifier les biais | Outils comme Debiasing Datasets de Google |
| **Débogage des données** | Supprimer ou corriger les biais identifiés | Supprimer les corrélations genrées des données |
| **Fine-tuning éthique** | Ajuster le modèle avec des données équilibrées | Fine-tuner sur des datasets incluant des minorités |
| **Contraintes explicites** | Ajouter des régularisations pendant l'entraînement | Pénaliser les associations discriminantes |
| **Équipes diversifiées** | Impliquer des profils variés dans le développement | Des équipes multiculturelles identifient plus de biais |

**Exemple pratique :** Google a développé des outils de débiasing comme le **What-If Tool** pour visualiser et corriger les biais dans les modèles.

---

### B. Pour la désinformation et les hallucinations

| Stratégie | Description | Exemple |
|:---|:---|:---|
| **Fact-checking intégré** | Vérifier les affirmations du modèle | Connecter le LLM à des bases de connaissances factuelles |
| **Transparence des sources** | Citer les sources des informations | Réponses avec références vérifiables |
| **Détection des hallucinations** | Classifier les réponses par niveau de confiance | Afficher un score de confiance |
| **Human-in-the-loop** | Validation humaine pour les cas critiques | Les réponses importantes sont revues |

## 1️⃣ Auto-attention et Attention multi-têtes

### A. Mécanisme d'auto-attention (Self-Attention)

#### 🔍 Définition et fonctionnement

L'auto-attention est le mécanisme fondamental qui permet aux Transformers de capturer les dépendances entre tous les mots d'une séquence, quelle que soit leur distance.

#### 📐 Formule mathématique

$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V$$

Où :
- **Q (Query)** : Représente ce que chaque token "cherche"
- **K (Key)** : Représente ce que chaque token "offre"
- **V (Value)** : Représente l'information réelle de chaque token
- **$d_k$** : Dimension des clés (facteur de scaling)

#### 🔄 Étapes détaillées du fonctionnement

| Étape | Description | Exemple concret |
|:---|:---|:---|
| **1. Création des matrices** | Pour chaque token, le modèle crée 3 vecteurs : Q, K, V en multipliant l'embedding par des matrices de poids | Le mot "chat" → Q_chat, K_chat, V_chat |
| **2. Calcul des scores** | Chaque Query est comparée à toutes les Keys via un produit scalaire | Q_chat · K_souris = score de similarité |
| **3. Scaling** | Division par $\sqrt{d_k}$ pour stabiliser les gradients | Score / √64 pour une dimension 64 |
| **4. Softmax** | Normalisation des scores en probabilités (somme = 1) | Attention distribution: [0.7, 0.2, 0.1] |
| **5. Pondération** | Chaque Value est multipliée par son poids d'attention | 0.7 × V_souris + 0.2 × V_chat + ... |
| **6. Somme** | Toutes les valeurs pondérées sont additionnées | Nouvelle représentation du token "chat" |

#### 💡 Analogie intuitive

**Métaphore de la bibliothèque :**
- **Query (Q)** = Votre question à un bibliothécaire
- **Key (K)** = L'index des livres dans la bibliothèque
- **Value (V)** = Le contenu réel des livres

Le bibliothécaire (le modèle) compare votre question (Q) à tous les index de livres (K) pour trouver les livres les plus pertinents, puis vous donne leur contenu (V) pondéré par leur pertinence.

#### 🎯 Avantages de l'auto-attention

| Avantage | Explication |
|:---|:---|
| **Dépendances à longue portée** | Capture des relations entre des mots éloignés dans la phrase |
| **Parallélisation** | Tous les tokens sont traités simultanément (pas séquentiel) |
| **Attention dynamique** | L'importance de chaque mot change selon le contexte |
| **Interprétabilité** | Les poids d'attention montrent ce que le modèle "regarde" |

---

### B. Attention multi-têtes (Multi-Head Attention)

#### 🎯 Objectif et avantages

| Aspect | Single-Head Attention | Multi-Head Attention |
|:---|:---|:---|
| **Nombre de têtes** | 1 seule | Plusieurs (typiquement 8-16) |
| **Perspectives** | Une seule façon de voir les relations | Perspectives multiples et diversifiées |
| **Capacité** | Limitée à un type de relation | Capture différents types de relations simultanément |
| **Flexibilité** | Moins flexible | Plus flexible et expressif |
| **Performance** | Performances de base | Performance supérieure |

#### 🔧 Fonctionnement

$$\text{MultiHead}(Q, K, V) = \text{Concat}(\text{head}_1, ..., \text{head}_h)W^O$$

Où : $\text{head}_i = \text{Attention}(QW_i^Q, KW_i^K, VW_i^V)$

#### 📊 Ce que chaque tête peut apprendre

| Type de relation | Exemple | Tête spécialisée |
|:---|:---|:---|
| **Relations syntaxiques** | Sujet-Verbe, Nom-Adjectif | Tête 1: Capture la structure grammaticale |
| **Relations sémantiques** | Synonymes, Hyperonymes | Tête 2: Comprend le sens des mots |
| **Dépendances à distance** | Références pronominales | Tête 3: Lie "il" à "Jean" |
| **Relations locales** | Contextes proches | Tête 4: Capture les bigrammes |
| **Relations globales** | Thème général | Tête 5: Comprend le sujet global |
| **Relations de position** | Ordre des mots | Tête 6: Capture la séquence |

#### 🌟 Exemple concret avec phrase

**Phrase :** *"Le scientifique a présenté ses recherches révolutionnaires à la conférence."*

| Tête d'attention | Relation capturée | Poids élevé entre... | Explication |
|:---|:---|:---|:---|
| **Tête 1 (Syntaxique)** | Sujet-Verbe | "scientifique" ↔ "a présenté" | Qui a présenté ? |
| **Tête 2 (Possession)** | Possessif | "ses" ↔ "recherches" | Les recherches de qui ? |
| **Tête 3 (Attribut)** | Qualificatif | "recherches" ↔ "révolutionnaires" | Quelles recherches ? |
| **Tête 4 (Lieu)** | Relation spatiale | "présenté" ↔ "conférence" | Où a-t-il présenté ? |
| **Tête 5 (Agent)** | Rôle sémantique | "scientifique" ↔ "conférence" | Qui participe à la conférence ? |
| **Tête 6 (Co-référence)** | Référence | "Le scientifique" ↔ "a" | Accord sujet-verbe |
| **Tête 7 (Temps)** | Temps verbal | "a présenté" ↔ contexte | Temps du verbe |
| **Tête 8 (Thème)** | Sujet principal | "scientifique" ↔ "recherches" | Thème central de la phrase |

**Résultat :** L'attention multi-têtes permet au modèle de comprendre simultanément :
1. Qui a fait quoi (sujet-action)
2. Quelles sont les caractéristiques de l'action
3. Où et dans quel contexte cela s'est passé
4. Les relations grammaticales et sémantiques complexes

#### 📈 Avantages de l'attention multi-têtes

| Avantage | Explication |
|:---|:---|
| **Parallélisme des relations** | Plusieurs relations analysées en même temps |
| **Richesse des représentations** | Encodage plus riche et nuancé du langage |
| **Robustesse** | Moins sensible aux erreurs d'une tête spécifique |
| **Transfer learning** | Différentes têtes peuvent se spécialiser dans différentes tâches |

#### 🔧 Fonctionnement

| Caractéristique | Description |
|:---|:---|

| **Méthode** | Auto-régressive : chaque prédiction dépend des prédictions précédentes |
| **Entraînement** | Prédire le token t+1 en voyant les tokens 1 à t |

3. Voir "Le chat noir" → prédire "dort"

## 🌟 Exercise 6: Softmax Temperature: The Randomness Regulator


## 1️⃣ Scénarios de température

### A. Comprendre la température Softmax

La température est un paramètre qui contrôle le niveau de "créativité" ou "aléatoire" dans la génération de texte par un modèle de langage. Elle modifie la distribution de probabilité des tokens avant la sélection.

**Formule :**

$$P_i = \frac{\exp(z_i / T)}{\sum_j \exp(z_j / T)}$$

Où :
- $z_i$ = logits (scores bruts) du token i
- $T$ = température (paramètre de scaling)
- $P_i$ = probabilité finale du token i

**Effet de la température :**

| Température | Effet sur la distribution | Comportement du modèle |
|:---|:---|:---|
| **T < 1** | Distribution plus "pointue" (peakier) | Modèle plus déterministe, sélectionne les mots les plus probables |
| **T = 1** | Distribution normale (inchangée) | Comportement standard du modèle |
| **T > 1** | Distribution plus "plate" (flatter) | Modèle plus créatif, sélectionne des mots moins probables |

---

## 2️⃣ Conception d'application

### A. Histoires personnalisées pour enfants (Régulation de la créativité et de la cohérence)

#### 🎯 Objectif principal
Trouver le **juste équilibre** entre des histoires suffisamment **créatives** pour captiver l'attention des enfants, mais suffisamment **cohérentes** pour avoir un sens et être compréhensibles.

#### 📊 Stratégie de température adaptative

| Étape | Température | Raison | Exemple |
|:---|:---|:---|:---|
| **1. Introduction** | T = 0.7 | Créer un début captivant mais clair | *"Il était une fois, dans un petit village..."* |
| **2. Développement** | T = 0.8 - 0.9 | Ajouter des péripéties créatives | *"Soudain, un dragon apparaît..."* |
| **3. Climax** | T = 0.9 - 1.0 | Moments d'action et de surprise | *"Le dragon crache du feu mais..."* |
| **4. Résolution** | T = 0.6 - 0.7 | Conclusion claire et satisfaisante | *"Et ils vécurent heureux..."* |
| **5. Morale** | T = 0.4 - 0.5 | Message éducatif cohérent | *"La leçon est qu'il faut être courageux..."* |

#### 💡 Stratégies avancées

| Stratégie | Description | Exemple d'implémentation |
|:---|:---|:---|
| **Température variable** | Ajuster T selon l'âge de l'enfant | 3-5 ans: T=0.6, 6-8 ans: T=0.8, 9-12 ans: T=1.0 |
| **Top-p sampling** | Filtrer les tokens trop improbables | Garder 90% de la masse probabiliste |
| **Repetition penalty** | Éviter les répétitions | Pénaliser les mots déjà utilisés |
| **Longueur adaptative** | Ajuster selon l'attention | Courte (T élevé pour captiver) vs Longue (T bas pour cohérence) |
| **Validation sémantique** | Vérifier la cohérence | Filtrer les éléments incohérents |

#### 🛠️ Implémentation pratique

```python
import random

def generer_histoire(age, theme, longueur="moyenne"):
    """
    Génère une histoire personnalisée avec température adaptative
    """
    
    # Ajuster la température selon l'âge
    if age <= 5:
        temperature = 0.6  # Simple et cohérent
    elif age <= 8:
        temperature = 0.8  # Plus créatif
    else:
        temperature = 1.0  # Très créatif pour les grands
    
    # Ajuster selon la longueur
    if longueur == "courte":
        temperature *= 1.1  # Plus créatif pour captiver
    elif longueur == "longue":
        temperature *= 0.9  # Plus cohérent pour tenir
    
    # Générer l'histoire avec des températures variables
    phases = [
        {"label": "Début", "T": temperature * 0.9},
        {"label": "Péripéties", "T": temperature * 1.1},
        {"label": "Climax", "T": temperature * 1.2},
        {"label": "Résolution", "T": temperature * 0.8},
    ]
    
    print(f"📖 Génération d'une histoire pour un enfant de {age} ans")
    print(f"Thème: {theme}")
    print(f"Température de base: {temperature:.2f}")
    print("\n📝 Phases de l'histoire:")
    for phase in phases:
        print(f"  • {phase['label']}: T={phase['T']:.2f}")
    
    return "Histoire générée avec succès!"

### (Optional) Quick Generation Demo Across Temperatures
Note: Requires `transformers` and model download; skip if offline.


````python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_name = "gpt2"  # TODO: choose a causal LM
prompt = "Artificial intelligence will"  # TODO: write your own prompt

pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(model_name),
    tokenizer=AutoTokenizer.from_pretrained(model_name),
    device=0 if torch.cuda.is_available() else -1,
)

for T in [0.2, 1.0, 1.5]:
    out = pipe(prompt, max_new_tokens=40, temperature=T, do_sample=True)
    print("--- Temperature:", T)
    print(out[0]["generated_text"])  # Inspect how style changes with T
````
