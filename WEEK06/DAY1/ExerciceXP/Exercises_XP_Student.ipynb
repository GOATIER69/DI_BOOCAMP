## 🌟 Exercise 1 · What are Large Language Models?


### 1.1 Define LLMs
Les grands modèles de langage (LLM) sont des systèmes d'intelligence artificielle entraînés sur d'énormes quantités de données textuelles (livres, articles, sites web, etc.). Ils apprennent à comprendre et à générer du langage humain en identifiant des motifs, des relations et des structures dans le texte.

À quoi ils servent :

Génération de texte (rédaction, création de contenu)

Traduction automatique

Résumé de documents

Réponse à des questions

Analyse de sentiment

Code et programmation

Assistance conversationnelle (chatbots)

Et bien d'autres applications basées sur le langage naturel

## 🌟 Exercise 2 · Transformer Architecture and Tokenization


La tokenisation est le processus qui consiste à découper un texte en unités plus petites appelées "tokens". C'est comme décomposer une phrase en morceaux compréhensibles par le modèle.

```python
text = None  # TODO: provide a sentence of your choice
if text is None:
    raise ValueError("Define the variable `text` with a short sentence.")

# 2. Tokenize input text
text = "Le machine learning est fascinant et transforme notre monde."  # Phrase de votre choix

tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text, add_special_tokens=True)

print(f"Original Text: {text}")
print(f"Tokens: {tokens}")
print(f"Token IDs: {token_ids}")

# 3. Visualizing the tokenization process
plt.figure(figsize=(10, 4))
plt.bar(tokens, token_ids, color="skyblue")
plt.xlabel("Tokens")
plt.ylabel("Token IDs")
plt.title("Tokenization Visualization: Tokens vs Token IDs")
plt.xticks(rotation=45, ha='right')  # Rotation pour meilleure lisibilité
plt.tight_layout()
plt.show()
```

Le Pré-entraînement (Pre-training)
Définition avec mes propres mots :
C'est la phase où le modèle apprend les bases du langage sur une quantité massive de données non labellisées (internet, livres, articles, etc.). C'est comme apprendre à lire, écrire et comprendre les structures grammaticales avant de se spécialiser.

La Mise au point (Fine-tuning)
Définition avec mes propres mots :
C'est la phase où l'on spécialise le modèle pré-entraîné sur une tâche spécifique en utilisant des données labellisées. C'est comme après les études générales, on suit une formation professionnelle pour un métier particulier.

## 🌟 Exercice 5

```python
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device=0 if torch.cuda.is_available() else -1,
)

# Création d'une nouvelle phrase d'entrée
input_text = "L'intelligence artificielle va transformer"  # Phrase de votre choix

# Tokenisation de l'entrée
input_ids = tokenizer.encode(input_text, return_tensors="pt")

# Génération du texte
with torch.no_grad():
    output_ids = model.generate(
        input_ids,
        max_length=50,  # Longueur maximale du texte généré
        num_return_sequences=1,  # Nombre de séquences à générer
        temperature=0.7,  # Contrôle la créativité (0 = déterministe, plus haut = plus créatif)
        do_sample=True,  # Utiliser l'échantillonnage pour plus de diversité
        pad_token_id=tokenizer.eos_token_id  # Éviter les avertissements
    )

# Décodage du texte généré
output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)

print(f"Input: {input_text}")
print(f"Generated Output: {output_text}")
```
