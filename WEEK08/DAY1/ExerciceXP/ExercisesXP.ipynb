4) Write a 3?5 sentence comparative paragraph.  
5) Answer the healthcare prompt (1?2 sentences) about retraining on clinical data.

**Deliverables:** paragraph + healthcare answer.

```

# Definitions and key characteristics for each openness level
open_source_levels = {
    "Fully Open": {
        "definition": "Every component is released: architecture, pretrained weights, training code, and often the training data or a detailed recipe for it.",
        "what_is_open": ["Model architecture and code", "Pretrained + fine-tuned weights", "Training/data pipeline details (varies by project)"],
        "what_you_can_do": ["Inspect every internal component", "Fully retrain or modify the model from scratch", "Reproduce or audit the training process"],
        "what_you_cannot_do": ["N/A in principle, though license terms (e.g. attribution, use restrictions) may still apply"],
    },
    "Weights Released": {
        "definition": "The pretrained model weights are published (often with a model card and some usage terms), but the full training code, dataset, and pipeline are not shared.",
        "what_is_open": ["Pretrained weights", "Basic architecture description (usually enough to load and run the model)"],
        "what_you_can_do": ["Run inference locally", "Fine-tune on your own data (LoRA, full fine-tune, etc.)"],
        "what_you_cannot_do": ["Reproduce the original training run", "Fully audit how the base model was trained or what data it saw"],
    },
    "Architecture Only": {
        "definition": "Only the model's code/architecture (and sometimes a paper) is published; no pretrained weights are released.",
        "what_is_open": ["Model architecture/code", "Research paper describing design choices"],
        "what_you_can_do": ["Study and reimplement the architecture", "Train your own version from scratch if you have the compute/data"],
        "what_you_cannot_do": ["Use the original pretrained model directly", "Skip the (often very expensive) pretraining step"],
    },
}
open_source_levels
```

```

# Comparison table (keep markdown formatting)
comparison_table = """| Openness level | What's open? | Impact on retraining/modifying |
| --- | --- | --- |
| Fully Open | Architecture, code, weights, and (often) training data/recipe | You can inspect, modify, and retrain the entire model from scratch |
| Weights Released | Only the pretrained weights (plus enough architecture info to run them) | You can fine-tune on top of the weights, but can't reproduce or fully audit the original training |
| Architecture Only | Just the model design/code, no pretrained weights | You know the structure but must pretrain from zero yourself to get a usable model |"""
print(comparison_table)
```

```

# 3-5 sentence comparative paragraph
comparative_paragraph = """Fully open models give the most freedom: the architecture, weights, and often the training pipeline are all public, so anyone can inspect, retrain, or repurpose the model without restriction. Weights-released models sit in the middle - you get usable, pretrained weights that you can fine-tune for your own domain, but you never see the original training code or data, so you can't reproduce the base model or fully verify how it was built. Architecture-only releases are the most limited in practice: you understand exactly how the model is built, but without pretrained weights you have to pay the full cost of pretraining before you have anything usable. The main trade-off across all three levels is transparency and control versus cost and convenience: fully open models maximize control and auditability, while architecture-only releases maximize transparency of design at the expense of a ready-to-use model."""
print(comparative_paragraph)
```

```

# 1-2 sentence healthcare prompt answer
healthcare_prompt_answer = """A Fully Open model is essential for a healthcare assistant that must be retrained on clinical data, because full access to the weights, architecture, and training pipeline lets the team fine-tune and validate the model thoroughly and audit exactly how patient data influences its behavior. This level of transparency is critical for compliance (e.g. HIPAA/GDPR-style requirements) and for demonstrating to regulators and clinicians that the model's outputs can be traced and trusted."""
print(healthcare_prompt_answer)
```
